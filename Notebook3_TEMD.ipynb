## **<font color="navy">Inferencia Estadística</font>**

La **<font color="crimson">inferencia estadística</font>** es el conjunto de métodos que permiten obtener conclusiones sobre una población a partir de una muestra.

En la práctica, no siempre es posible analizar toda la población, por lo que se trabaja con muestras para estimar parámetros desconocidos.


### **<font color="darkgreen">Definición</font>**

Sea $X$ una variable aleatoria continua.

Se tiene una muestra aleatoria:

$$
X_1, X_2, \dots, X_n
$$

Se dice que es una muestra de una población si:

$$
X_1, X_2, \dots, X_n \sim F(\theta)
$$

donde:

- $F(\theta)$ es la distribución de la población  
- $\theta$ es un parámetro desconocido  


### **<font color="darkgreen">Objetivo de la Inferencia</font>**

El objetivo es estimar el parámetro $\theta$ utilizando la información contenida en la muestra.

Es decir, construir una función:

$$
\hat{\theta} = g(X_1, X_2, \dots, X_n)
$$

que permita aproximar el valor real del parámetro.

En inferencia estadística, es fundamental distinguir entre población y muestra.

- **Población**: conjunto total de elementos de interés  
- **Muestra**: subconjunto de la población utilizado para el análisis  

### **<font color="darkgreen">Representación Formal</font>**

Sea $X$ una variable aleatoria que describe a la población:

$$
X \sim F(\theta)
$$

donde $\theta$ es un parámetro desconocido.

Se toma una muestra aleatoria:

$$
X_1, X_2, \dots, X_n
$$

tal que:

$$
X_i \sim F(\theta), \quad i = 1,2,\dots,n
$$

y además:

$$
X_1, X_2, \dots, X_n \text{ son independientes}
$$


La muestra contiene información sobre la población, por lo que puede utilizarse para estimar sus parámetros.

Sin embargo, al trabajar con una muestra se introduce incertidumbre, ya que no se observa toda la población.


##### **<font color="black">Ejemplo: </font>**

Supongamos que se desea conocer el ingreso promedio de todos los asegurados de una compañía.

- Población: todos los asegurados  
- Muestra: un subconjunto de asegurados seleccionados  

El objetivo es estimar:

$$
\mu = \mathbb{E}[X]
$$

utilizando la información de la muestra.

## **<font color="navy">Parámetros y Estadísticos</font>**

En inferencia estadística es fundamental distinguir entre **parámetros** y **estadísticos**, ya que toda la teoría se basa en la relación entre ambos.

La idea central es que los parámetros describen la población, mientras que los estadísticos se obtienen a partir de la muestra y se utilizan para inferir dichos parámetros.


## **<font color="darkgreen">Parámetros</font>**

Un **<font color="crimson">parámetro</font>** es una característica numérica que describe completamente a una población o distribución.

Generalmente es **desconocido** y se representa con letras griegas.

Sea:

$$
X \sim F(\theta)
$$

donde $\theta$ representa uno o varios parámetros poblacionales.


##### **<font color="black">Ejemplos</font>**

$$
\mu = \mathbb{E}[X] \quad \text{(media poblacional)}
$$

$$
\sigma^2 = \mathbb{V}[X] \quad \text{(varianza poblacional)}
$$

$$
\sigma = \sqrt{\mathbb{V}[X]} \quad \text{(desviacion estandar)}
$$

Los parámetros:

- Son constantes (no cambian)
- No se conocen en la práctica
- Definen completamente la distribución de la población



## **<font color="darkgreen">Estadísticos</font>**

Un **<font color="crimson">estadístico</font>** es una función de los datos muestrales, es decir, depende de:

$$
X_1, X_2, \dots, X_n
$$

A diferencia de los parámetros, los estadísticos **sí pueden calcularse**.


##### **<font color="black">Ejemplos</font>**

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i \quad \text{(media muestral)}
$$

$$
S^2 = \frac{1}{n-1} \sum_{i=1}^{n} (X_i - \bar{X})^2 \quad \text{(varianza muestral)}
$$


Los estadísticos:

- Son variables aleatorias (dependen de la muestra)
- Cambian de muestra a muestra
- Tienen su propia distribución (distribución muestral)


#### **<font color="darkgreen">Idea clave: </font>**

El objetivo de la inferencia es utilizar estadísticos para aproximar parámetros desconocidos:

$$
\bar{X} \approx \mu
$$

$$
S^2 \approx \sigma^2
$$

Sin embargo, esta aproximación introduce incertidumbre debido a la variabilidad muestral.

## **<font color="navy">Estimadores</font>**

Un **<font color="crimson">estimador</font>** es una función de la muestra utilizada para aproximar un parámetro poblacional.

Sea:

$$
\hat{\theta} = g(X_1, X_2, \dots, X_n)
$$

donde:

- $\theta$ es el parámetro poblacional  
- $\hat{\theta}$ es su estimador  


### **<font color="darkgreen">Estimación Puntual</font>**

La estimación puntual consiste en asignar un único valor al parámetro:

$$
\theta \approx \hat{\theta}
$$


##### **<font color="black">Ejemplo</font>**

$$
\hat{\mu} = \bar{X}
$$

En este caso, la media muestral se utiliza como estimador de la media poblacional.


### **<font color="darkgreen">Propiedades de los Estimadores</font>**

#### **1. Insesgadez**

Un estimador es insesgado si su valor esperado coincide con el parámetro:

$$
\mathbb{E}[\hat{\theta}] = \theta
$$

Si no se cumple, existe un sesgo:

$$
\text{Sesgo}(\hat{\theta}) = \mathbb{E}[\hat{\theta}] - \theta
$$



#### **2. Consistencia**

Un estimador es consistente si converge al valor real cuando el tamaño de muestra crece:

$$
\hat{\theta} \xrightarrow{p} \theta \quad \text{cuando } n \to \infty
$$


#### **3. Eficiencia**

Un estimador es eficiente si tiene la menor varianza posible entre los estimadores insesgados:

$$
\mathbb{V}[\hat{\theta}] \text{ minima}
$$


### **<font color="darkgreen">Error del Estimador</font>**

El error de estimación se define como:

$$
\hat{\theta} - \theta
$$

Este error es aleatorio y depende de la muestra seleccionada.


### **<font color="darkgreen">Error Cuadrático Medio</font>**

El desempeño de un estimador se evalúa mediante:

$$
\text{ECM}(\hat{\theta}) = \mathbb{E}[(\hat{\theta} - \theta)^2]
$$


#### **<font color="darkgreen">Descomposición del ECM</font>**

$$
\text{ECM}(\hat{\theta}) = \mathbb{V}[\hat{\theta}] + \left(\text{Sesgo}(\hat{\theta})\right)^2
$$

El ECM combina:

- Variabilidad del estimador  
- Error sistemático (sesgo)  


#### **<font color="darkgreen">En general: </font>**

Un buen estimador debe:

- Tener sesgo cercano a cero  
- Tener baja varianza  
- Ser consistente  

## **<font color="navy">Métodos de Estimación</font>**

Hasta ahora, se han definido las propiedades que debe tener un buen estimador, como insesgadez, consistencia y eficiencia.

Sin embargo, surge una pregunta fundamental:

**¿Cómo se construye un estimador $\hat{\theta}$ a partir de los datos?**


Existen diferentes métodos para obtener estimadores, entre los más importantes:

- **Método de momentos**  
- **Máxima verosimilitud (MLE)**

Cada método propone una forma distinta de utilizar la información de la muestra.

Recordemos que:

$$
\mathbb{E}[X] = \mu
$$

y que en la muestra:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

es una aproximación natural de la media poblacional.

Si los momentos poblacionales describen la distribución, entonces podemos estimar los parámetros igualando:

- Momentos poblacionales  
- Momentos muestrales  

Esto da lugar al **<font color="crimson">método de momentos</font>**, el cual consiste en igualar los momentos teóricos con los momentos empíricos para obtener estimadores.

## **<font color="navy">Método de los Momentos</font>**

El **<font color="crimson">Método de Momentos</font>** es una técnica de estimación que permite obtener estimadores de parámetros poblacionales utilizando los momentos de la distribución.


### **<font color="darkgreen">Momentos Poblacionales</font>**

Sea $X$ una variable aleatoria.

El momento de orden $k$ respecto al origen se define como:

$$
\mu'_k = \mathbb{E}[X^k]
$$

Estos momentos caracterizan la forma de la distribución.

### **<font color="darkgreen">Momentos Muestrales</font>**

Dada una muestra:

$$
X_1, X_2, \dots, X_n
$$

el momento muestral de orden $k$ se define como:

$$
m_k = \frac{1}{n} \sum_{i=1}^{n} X_i^k
$$

**De esta manera:**

Los momentos poblacionales contienen información sobre los parámetros desconocidos $\theta$.

Por otro lado, los momentos muestrales son cantidades observables que aproximan a los momentos poblacionales.

Esto sugiere la siguiente aproximación:

$$
\mathbb{E}[X^k] \approx \frac{1}{n} \sum_{i=1}^{n} X_i^k
$$


### **<font color="darkgreen">Principio del Método</font>**

El método de los momentos parte de una idea muy simple pero poderosa:
los **momentos poblacionales** contienen información sobre los parámetros desconocidos, mientras que los **momentos muestrales** son cantidades que sí podemos calcular.

Por lo tanto, se propone aproximar:

$$
\mu'_k = \mathbb{E}[X^k] \approx m_k = \frac{1}{n} \sum_{i=1}^{n} X_i^k
$$


- $\mu'_k$ → lo que dice la teoría (población)  
- $m_k$ → lo que dicen los datos (muestra)  

La idea es:

> "Hacer que el modelo se parezca lo más posible a los datos igualando sus momentos"

Se igualan los momentos teóricos con los muestrales:

$$
\mu'_k = m_k
$$

para:

$$
k = 1, 2, \dots, p
$$

donde $p$ es el número de parámetros desconocidos.

**¿Por qué hasta $p$ momentos?**

Porque:

- Cada ecuación aporta información  
- Necesitamos tantas ecuaciones como parámetros  

Si hay $p$ parámetros → necesitamos $p$ ecuaciones para poder resolver el sistema.


### **<font color="darkgreen">Construcción del Estimador</font>**

Sea:

$$
\theta = (\theta_1, \theta_2, \dots, \theta_p)
$$

un vector de parámetros.

Los momentos poblacionales dependen de $\theta$, por ejemplo:

$$
\mathbb{E}[X] = f_1(\theta)
$$

$$
\mathbb{E}[X^2] = f_2(\theta)
$$

$$
\vdots
$$

$$
\mathbb{E}[X^p] = f_p(\theta)
$$


### **<font color="darkgreen">Sistema de Ecuaciones</font>**

Se iguala cada momento:

$$
f_1(\theta) = m_1
$$

$$
f_2(\theta) = m_2
$$

$$
\vdots
$$

$$
f_p(\theta) = m_p
$$


Este sistema se resuelve algebraicamente para encontrar:

$$
\hat{\theta}_{MM}
$$


### **<font color="darkgreen">Relación con la Ley de los Grandes Números</font>**

Por la Ley de los Grandes Números:

$$
m_k \xrightarrow{p} \mathbb{E}[X^k]
$$


### **<font color="darkgreen">Unicidad</font>**

El sistema puede tener:

- **Solución única** → caso ideal  
- **Múltiples soluciones** → se requiere criterio adicional  
- **Sin solución válida** → el método falla  

Esto depende de:

- La forma de la distribución  
- La relación entre momentos y parámetros  


### **<font color="darkgreen">Limitaciones</font>**

Aunque es intuitivo, el método tiene limitaciones:

- No garantiza mínima varianza  
- Puede generar estimadores sesgados  
- No siempre produce estimadores eficientes  

## **<font color="purple">Ejemplo: </font>**

Sea:

$$
X_1, X_2, \dots, X_n \sim \text{Exp}(\lambda)
$$

una muestra aleatoria de una distribución exponencial, donde $\lambda > 0$ es un parámetro desconocido.

#### **<font color="black">Calcular:</font>**

1. El estimador del parámetro $\lambda$ mediante el método de los momentos  


## **<font color="navy">Solución</font>**


Sabemos que el primer momento poblacional es:

$$
\mathbb{E}[X] = \frac{1}{\lambda}
$$

Y el primer momento muestral es:

$$
m_1 = \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$


Aplicando el método de los momentos, es decir, igualamos el momento poblacional con el muestral:

$$
\mathbb{E}[X] = m_1
$$

$$
\frac{1}{\lambda} = \bar{X}
$$


Despejando $\lambda$:

$$
\lambda = \frac{1}{\bar{X}}
$$

Por lo tanto el estimador es:

$$
\hat{\lambda} = \frac{1}{\bar{X}}
$$


- Si la media muestral es grande → $\lambda$ es pequeño  
- Si la media muestral es pequeña → $\lambda$ es grande  


## **<font color="purple">Ejemplo: </font>**

Sea:

$$
X_1, X_2, \dots, X_n \sim \text{Gamma}(\alpha, \beta)
$$

una muestra aleatoria, donde $\alpha > 0$ (forma) y $\beta > 0$ (escala) son parámetros desconocidos.


#### **<font color="black">Calcular:</font>**

1. Los estimadores de $\alpha$ y $\beta$ mediante el método de los momentos  


### **<font color="navy">Solución</font>**

Para la distribución Gamma sabemos que los momentos poblacionales son:

$$
\mathbb{E}[X] = \alpha \beta
$$

$$
\mathbb{E}[X^2] = \alpha(\alpha + 1)\beta^2
$$

Y sabemos que los momentos muestrales son:

Primer momento:

$$
m_1 = \bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

Segundo momento:

$$
m_2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2
$$

Aplicando el método de los momentos, es decir, igualamos los momentos poblacionales con los muestrales:


$$
\alpha \beta = \bar{X}
$$

$$
\alpha(\alpha + 1)\beta^2 = m_2
$$

Ahora procedemos a resolver el sistema de ecuaciones:

De la primera ecuación:

$$
\beta = \frac{\bar{X}}{\alpha}
$$

Sustituyendo en la segunda:

$$
\alpha(\alpha + 1)\left(\frac{\bar{X}}{\alpha}\right)^2 = m_2
$$

Simplificando:

$$
\frac{(\alpha + 1)\bar{X}^2}{\alpha} = m_2
$$

Despejando:

$$
\alpha (m_2 - \bar{X}^2) = \bar{X}^2
$$

Los estimadores son:

$$
\hat{\alpha} = \frac{\bar{X}^2}{m_2 - \bar{X}^2}
$$

$$
\hat{\beta} = \frac{m_2 - \bar{X}^2}{\bar{X}}
$$

Sustituyendo los valores de $\bar{X}$ y $m_2$

Recordando que:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

$$
m_2 = \frac{1}{n} \sum_{i=1}^{n} X_i^2
$$

**Estimador de $\alpha$**

$$
\hat{\alpha} =
\frac{\left(\frac{1}{n} \sum_{i=1}^{n} X_i \right)^2}
{\frac{1}{n} \sum_{i=1}^{n} X_i^2 - \left(\frac{1}{n} \sum_{i=1}^{n} X_i \right)^2}
$$

**Estimador de $\beta$**

$$
\hat{\beta} =
\frac{\frac{1}{n} \sum_{i=1}^{n} X_i^2 - \left(\frac{1}{n} \sum_{i=1}^{n} X_i \right)^2}
{\frac{1}{n} \sum_{i=1}^{n} X_i}
$$


- $\sum X_i$ → suma de los datos  
- $\sum X_i^2$ → suma de los datos al cuadrado  
- $\bar{X}$ → promedio  

Es decir, todo se calcula directamente con los datos observados.

Hasta ahora se ha estudiado el método de los momentos, el cual consiste en igualar los momentos teóricos con los momentos muestrales para obtener estimadores.

Sin embargo, surge una pregunta importante:

>**¿Existe un método que utilice directamente la distribución de los datos para estimar los parámetros?**


El método de **<font color="crimson">Máxima Verosimilitud</font>** propone estimar los parámetros como aquellos valores que hacen que los datos observados sean los más probables bajo el modelo.


A diferencia del método de los momentos, que se basa en igualar características de la distribución, la máxima verosimilitud se basa en maximizar la probabilidad (o densidad) de los datos observados.

## **<font color="navy">Método de Máxima Verosimilitud</font>**

El **<font color="crimson">Método de Máxima Verosimilitud (MLE)</font>** es una técnica fundamental de inferencia estadística que permite estimar parámetros poblacionales utilizando directamente la distribución de los datos.

A diferencia del método de los momentos, que iguala características (momentos), este método utiliza toda la información probabilística del modelo.


### **<font color="darkgreen">Planteamiento</font>**

Sea:

$$
X_1, X_2, \dots, X_n \sim f(x; \theta)
$$

una muestra aleatoria i.i.d., donde:

- $f(x; \theta)$ es la función de densidad (o masa)  
- $\theta$ es el parámetro desconocido  


Dado que las observaciones son independientes:

$$
f(X_1, X_2, \dots, X_n; \theta) = \prod_{i=1}^{n} f(X_i; \theta)
$$


### **<font color="darkgreen">Función de Verosimilitud</font>**

La función de verosimilitud se define como:

$$
L(\theta) = \prod_{i=1}^{n} f(X_i; \theta)
$$


- Matemáticamente es la misma función que la densidad conjunta  
- Conceptualmente se interpreta como función de $\theta$  
- Los datos $X_i$ se consideran fijos  


El estimador de máxima verosimilitud se define como el valor del parámetro que maximiza la función de verosimilitud:

$$
\hat{\theta}_{MLE} = \underset{\theta}{\arg\max} \; L(\theta)
$$

Es decir, se busca el valor de $\theta$ que hace más probable la muestra observada.


**¿Por qué maximizar?**

Porque queremos encontrar el parámetro que hace que:

> “los datos observados sean los más probables bajo el modelo”


### **<font color="darkgreen">Log-Verosimilitud</font>**

Para simplificar los cálculos, se utiliza el logaritmo de la función de verosimilitud:

$$
\ell(\theta) = \ln L(\theta)
$$


Recordando que:

$$
L(\theta) = \prod_{i=1}^{n} f(X_i; \theta)
$$

al aplicar logaritmo:

$$
\ell(\theta) = \ln \left( \prod_{i=1}^{n} f(X_i; \theta) \right)
$$

Usando propiedades de logaritmos:

$$
\ell(\theta) = \sum_{i=1}^{n} \ln f(X_i; \theta)
$$


La log-verosimilitud mide qué tan bien el modelo explica los datos, pero en una escala logarítmica.

- Valores más grandes → mejor ajuste  
- Valores más pequeños → peor ajuste  


#### **<font color="black">¿Por qué usar logaritmo?</font>**

El uso del logaritmo tiene varias ventajas clave:

##### **1. Simplificación algebraica**

Convierte productos en sumas:

$$
\prod \rightarrow \sum
$$

Esto facilita enormemente los cálculos.


##### **2. Facilita derivación**

Las derivadas de sumas son más sencillas que las de productos:

$$
\frac{d}{d\theta} \ell(\theta) = \sum_{i=1}^{n} \frac{d}{d\theta} \ln f(X_i; \theta)
$$


##### **3. No cambia el máximo**

El logaritmo es una función monótona creciente, por lo que:

$$
\arg\max_{\theta} L(\theta) = \arg\max_{\theta} \ell(\theta)
$$


### **<font color="darkgreen">Cálculo del Estimador</font>**

El estimador se obtiene resolviendo:

$$
\frac{d}{d\theta} \ell(\theta) = 0
$$


Para verificar que se trata de un máximo:

$$
\frac{d^2}{d\theta^2} \ell(\theta) < 0
$$


### **<font color="darkgreen">Caso Multivariado</font>**

Si $\theta = (\theta_1, \dots, \theta_p)$:

$$
\frac{\partial \ell}{\partial \theta_1} = 0, \quad
\frac{\partial \ell}{\partial \theta_2} = 0, \quad \dots
$$

Se obtiene un sistema de ecuaciones.


### **<font color="darkgreen">Propiedades Asintóticas</font>**

Bajo condiciones regulares, los estimadores MLE cumplen:

- **Consistencia**:
$$
\hat{\theta}_{MLE} \xrightarrow{p} \theta
$$

- **Normalidad asintótica**:
$$
\hat{\theta}_{MLE} \approx N\left(\theta, \frac{1}{I(\theta)}\right)
$$

donde $I(\theta)$ es la información de Fisher.


### **<font color="darkgreen">Información de Fisher</font>**

Se define como:

$$
I(\theta) = -\mathbb{E}\left[\frac{d^2}{d\theta^2} \ell(\theta)\right]
$$

Mide la cantidad de información que los datos contienen sobre $\theta$.

## **<font color="purple">Ejemplo:</font>**

Sea:

$$
X_1, X_2, \dots, X_n \sim \text{Exp}(\lambda)
$$

una muestra aleatoria de una distribución exponencial, donde $\lambda > 0$ es un parámetro desconocido.

La función de densidad está dada por:

$$
f(x;\lambda) = \lambda e^{-\lambda x}, \quad x > 0
$$


#### **<font color="black">Calcular:</font>**

El estimador de $\lambda$ mediante el método de máxima verosimilitud  


### **<font color="navy">Solución</font>**

Dado que las observaciones son independientes:

$$
L(\lambda) = \prod_{i=1}^{n} f(X_i; \lambda)
$$

Sustituyendo la densidad:

$$
L(\lambda) = \prod_{i=1}^{n} \lambda e^{-\lambda X_i}
$$

Simplificando:

$$
L(\lambda) = \lambda^n \exp\left(-\lambda \sum_{i=1}^{n} X_i\right)
$$

Aplicando logaritmo para la Log-verosimilitud:

$$
\ell(\lambda) = \ln L(\lambda)
$$

$$
\ell(\lambda) = n \ln \lambda - \lambda \sum_{i=1}^{n} X_i
$$

Derivando:

$$
\frac{d}{d\lambda} \ell(\lambda) = \frac{n}{\lambda} - \sum_{i=1}^{n} X_i
$$

Igualando a cero:

$$
\frac{n}{\lambda} - \sum_{i=1}^{n} X_i = 0
$$

Despejando:

$$
\frac{n}{\lambda} = \sum_{i=1}^{n} X_i
$$

$$
\lambda = \frac{n}{\sum_{i=1}^{n} X_i}
$$

Por tanto, el estimador por MLE es:

$$
\hat{\lambda} = \frac{n}{\sum_{i=1}^{n} X_i}
$$


Recordando que:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

entonces:

$$
\hat{\lambda} = \frac{1}{\bar{X}}
$$


> El estimador de máxima verosimilitud coincide con el estimador por método de momentos:

$$
\hat{\lambda}_{MLE} = \hat{\lambda}_{MM}
$$

Esto muestra que ambos métodos producen el mismo resultado en este caso.

## **<font color="purple">Ejemplo:</font>**

Sea:

$$
X_1, X_2, \dots, X_n \sim N(\mu, \sigma^2)
$$

una muestra aleatoria.



#### **<font color="black">Calcular:</font>**

Los estimadores de $\mu$ y $\sigma^2$ mediante máxima verosimilitud  


### **<font color="navy">Solución</font>**
La funcion de densidad de la distribucion normal esta dada por:

$$
f(x;\mu,\sigma^2) =
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{1}{2\sigma^2}(x-\mu)^2}
$$

La funcion de verosimilitud es:

$$
L(\mu,\sigma^2) =
\prod_{i=1}^{n} f(X_i;\mu,\sigma^2)
=
\prod_{i=1}^{n}
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{1}{2\sigma^2}(X_i-\mu)^2}
$$

Desarrollando el producto se tiene:

$$
=
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{1}{2\sigma^2}(X_1-\mu)^2}
\cdot
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{1}{2\sigma^2}(X_2-\mu)^2}
\cdots
\frac{1}{\sqrt{2\pi\sigma^2}}
e^{-\frac{1}{2\sigma^2}(X_n-\mu)^2}
$$

Simplificando:

$$
=
\left(\frac{1}{\sqrt{2\pi\sigma^2}}\right)^n
\cdot
e^{-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(X_i-\mu)^2}
$$

$$
=
(2\pi\sigma^2)^{-n/2}
\cdot
e^{-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(X_i-\mu)^2}
$$

Aplicando logaritno para la Log-verosimilitud:

$$
\ell(\mu,\sigma^2) = \ln L(\mu,\sigma^2)
$$

$$
=
-\frac{n}{2}\ln(2\pi\sigma^2)
-\frac{1}{2\sigma^2}\sum_{i=1}^{n}(X_i-\mu)^2
$$


Se calculan las derivadas respecto a ambos parametros:

**Respecto a $\mu$:**

$$
\frac{\partial \ell}{\partial \mu}
=
\frac{1}{\sigma^2}\sum_{i=1}^{n}(X_i-\mu)
$$

**Respecto a $\sigma^2$:**

$$
\frac{\partial \ell}{\partial \sigma^2}
=
-\frac{n}{2\sigma^2}
+
\frac{1}{2\sigma^4}
\sum_{i=1}^{n}(X_i-\mu)^2
$$

Por lo tanto, se tiene el sistema de ecuaciones:

$$
\frac{1}{\sigma^2}\sum_{i=1}^{n}(X_i-\mu) = 0
$$

$$
-\frac{n}{2\sigma^2}
+
\frac{1}{2\sigma^4}
\sum_{i=1}^{n}(X_i-\mu)^2 = 0
$$

**Resolviendo para $\mu$:**

$$
\sum_{i=1}^{n}(X_i-\mu)=0
$$

$$
\sum_{i=1}^{n}X_i - n\mu = 0
$$

$$
\mu = \frac{1}{n}\sum_{i=1}^{n}X_i
$$


**Resolviendo para $\sigma^2$:**

$$
-\frac{n}{2\sigma^2}
+
\frac{1}{2\sigma^4}
\sum_{i=1}^{n}(X_i-\mu)^2 = 0
$$

Multiplicando por $2\sigma^4$ para eliminar denominadores:

$$
- n\sigma^2 + \sum_{i=1}^{n}(X_i-\mu)^2 = 0
$$

Despejando:

$$
\sum_{i=1}^{n}(X_i-\mu)^2 = n\sigma^2
$$

$$
\sigma^2 = \frac{1}{n} \sum_{i=1}^{n}(X_i-\mu)^2
$$


Como ya se obtuvo que:

$$
\hat{\mu} = \bar{X}
$$

entonces:

$$
\hat{\sigma}^2 =
\frac{1}{n} \sum_{i=1}^{n}(X_i-\bar{X})^2
$$

## **<font color="purple">Ejemplo:</font>**

Se lanza una moneda $n = 100$ veces, obteniendo $x = 55$ éxitos (soles).

Sea:

$$
X \sim \text{Binomial}(n, p)
$$

donde $p$ es la probabilidad de éxito en un lanzamiento.

#### **<font color="black">Calcular:</font>**

El estimador de máxima verosimilitud (MLE) para $p$


### **<font color="navy">Solución</font>**

La Función de verosimilitud es:

$$
L(p) = \binom{100}{55} p^{55} (1-p)^{45}
$$


Como el coeficiente binomial no depende de $p$, se puede omitir:

$$
L(p) \propto p^{55} (1-p)^{45}
$$

Aplicando logaritmo para la Log-verosimilitud:

$$
\ell(p) = \ln L(p)
$$

$$
\ell(p) = 55 \ln p + 45 \ln(1-p)
$$

Derivando:

$$
\frac{d}{dp} \ell(p) = \frac{55}{p} - \frac{45}{1-p}
$$

Igualando a cero:

$$
\frac{55}{p} - \frac{45}{1-p} = 0
$$

Despejando:

$$
\frac{55}{p} = \frac{45}{1-p}
$$

$$
55(1-p) = 45p
$$

$$
55 - 55p = 45p
$$

$$
55 = 100p
$$

$$
\hat{p}_{MLE} = \frac{55}{100}
$$

**Por tanto el estimador es:**

$$
\hat{p} = 0.55
$$

```python
# Importamos la librería SymPy para matemáticas símbolica
import sympy as sp

# Definimos la variable simbólica p (probabilidad de éxito). Se restringe a valores positivos
p = sp.symbols('p', positive=True)

# Definimos los datos del problema
n = 100   # número de ensayos
x = 55    # número de éxitos observados

# Definimos la función de verosimilitud sin la constante binomial (ya que no depende de p y no afecta el máximo)
L = p**x * (1 - p)**(n - x)

# Calculamos la log-verosimilitud
logL = sp.log(L)

# Derivamos la log-verosimilitud respecto a p
dlogL = sp.diff(logL, p)

# Resolvemos la ecuación de primer orden (derivada = 0)
solutions = sp.solve(dlogL, p)

# Seleccionamos la solución válida (real y dentro del intervalo)
p_hat = solutions[0]

# Mostramos el resultado simbólico y numérico
print("Estimador MLE:", float(p_hat))
```

## **<font color="purple">Ejemplo:</font>**

Se simula una muestra de tamaño $n = 100$ de variables Bernoulli con probabilidad de éxito $p$.

Sea:

$$
X_1, X_2, \dots, X_n \sim \text{Bernoulli}(p)
$$

donde cada observación toma valores:

$$
X_i \in \{0,1\}
$$


#### **<font color="black">Calcular: </font>**

El estimador de máxima verosimilitud para $p$ a partir de los datos simulados.


### **<font color="navy">Solucion:</font>**


Para una sola observación:

$$
f(x_i; p) = p^{x_i}(1-p)^{1-x_i}
$$

Para toda la muestra (funcion de verosimilitud):

$$
L(p) = \prod_{i=1}^{n} p^{x_i}(1-p)^{1-x_i}
$$


Aplicando logaritmo para la Log-verosimilitud:

$$
\ell(p) = \ln L(p)
$$

$$
\ell(p) = \sum_{i=1}^{n} \left[ x_i \ln p + (1-x_i)\ln(1-p) \right]
$$


Separando términos:

$$
\ell(p) = \left(\sum_{i=1}^{n} x_i\right)\ln p
+ \left(n - \sum_{i=1}^{n} x_i\right)\ln(1-p)
$$

Derivando:

$$
\frac{d}{dp}\ell(p) =
\frac{\sum x_i}{p}
-
\frac{n - \sum x_i}{1-p}
$$

Igualando a cero:

$$
\frac{\sum x_i}{p} =
\frac{n - \sum x_i}{1-p}
$$

Despejando:

$$
\hat{p} =
\frac{1}{n} \sum_{i=1}^{n} x_i
$$


$$
\hat{p} = \bar{X}
$$

El estimador de máxima verosimilitud para $p$ es simplemente la proporción de éxitos observados en la muestra.

Es decir:

- Si hay muchos unos → $p$ grande  
- Si hay pocos unos → $p$ pequeño  



```python
# Importamos las librerías necesarias

# numpy se usa para manejo de arreglos y operaciones numéricas
import numpy as np

# scipy.stats.binom se usa para generar variables aleatorias
from scipy.stats import binom


# Definimos los parámetros del problema
n = 100        # tamaño de la muestra
p_real = 0.7   # probabilidad verdadera de éxito


# Generamos una muestra de variables Bernoulli
# n=1 indica que cada observación es 0 o 1
# size=100 genera 100 observaciones
muestra = binom.rvs(n=1, p=p_real, size=n)


# Mostramos la muestra generada
print("Muestra generada:")
print(muestra)


# Calculamos el número total de éxitos (cantidad de unos)
num_exitos = np.sum(muestra)


# Calculamos el estimador de máxima verosimilitud
# Para Bernoulli/Binomial: p̂ = x / n
p_hat = num_exitos / n


# Mostramos resultados
print("\nNúmero de éxitos:", num_exitos)
print("Estimador MLE (p̂):", p_hat)
print("Valor real de p:", p_real)


# Interpretación básica en consola
print("\nInterpretación:")
print("El estimador es la proporción de éxitos observados en la muestra.")
print("Puede diferir de p_real debido a la aleatoriedad.")
```

## **<font color="purple">Ejemplo: </font>**

El tiempo de retardo para pacientes con enfermedad coronaria se modela como una variable aleatoria continua.

Sea:

$$
X \sim \text{Exp}(\theta)
$$
donde $\theta > 0$ es el parámetro desconocido.

Se tienen los siguientes tiempos de retardo (en minutos):

$$
525, 719, 2880, 150, 30, 251, 45, 858, 15, 47, 90, 56, 68, 6, 189, 180, 60, 60, 294, 747
$$


#### **<font color="black">Calcular</font>**

1. Encontrar el estimador de $\theta$ por el método de momentos  
2. Calcular su valor con los datos observados  


### **<font color="navy">Solución</font>**

**Momento poblacional:**

Para la distribución exponencial:

$$
\mathbb{E}[X] = \frac{1}{\theta}
$$

**Momento muestral_**

$$
\mathbb{E}[X] \approx \bar{X}
$$

Igualando momentos

$$
\frac{1}{\theta} = \bar{X}
$$

Despejando

$$
\hat{\theta} = \frac{1}{\bar{X}}
$$

**Resultado:**

Si:

$$
\bar{X} = 363.5
$$

entonces:

$$
\hat{\theta} = \frac{1}{363.5} \approx 0.00275
$$


- $\theta$ representa la tasa de ocurrencia por minuto  
- El tiempo promedio de retardo es:

$$
\frac{1}{\hat{\theta}} \approx 363.5 \text{ minutos}
$$


```python
# numpy se usa para manejo de datos y cálculos numéricos
import numpy as np

# scipy.optimize se usa para estimar MLE por optimización
from scipy.optimize import minimize

# Datos observados (tiempos de retardo en minutos)
datos = np.array([525, 719, 2880, 150, 30, 251, 45, 858, 15, 47, 90, 56, 68, 6, 189, 180, 60, 60, 294, 747], dtype=float)

# Tamaño de la muestra
n = len(datos)

# Media muestral
media_muestral = np.mean(datos)


# Método de Momentos
# Para Exp(θ): E[X] = 1/θ  →  θ = 1 / media
theta_mom = 1 / media_muestral


# Función de log-verosimilitud negativa
def neg_log_likelihood(theta, data):
    # Restricción: theta debe ser positivo
    if theta <= 0:
        return np.inf

    # Número de observaciones
    n = len(data)

    # Suma de los datos
    suma = np.sum(data)

    # Log-verosimilitud
    logL = n * np.log(theta) - theta * suma

    # Retornamos el negativo para minimizar
    return -logL


# Valor inicial
theta_inicial = 1 / media_muestral

# Optimización
resultado = minimize(
    neg_log_likelihood,
    x0=theta_inicial,
    args=(datos,),
    bounds=[(1e-10, None)]
)

# Estimador MLE
theta_mle = resultado.x[0]


# Resultados
print("Tamaño de la muestra:", n)
print("Media muestral:", media_muestral)

print("\nEstimador por Método de Momentos:", theta_mom)
print("Estimador por Máxima Verosimilitud:", theta_mle)

# Diferencia entre estimadores
print("\nDiferencia:", abs(theta_mom - theta_mle))


# Interpretación
print("\nTiempo promedio estimado:", media_muestral, "minutos")
print("Tasa estimada θ:", theta_mle, "eventos por minuto")
```

## **<font color="purple">Ejemplo: </font>**

Se observan las siguientes duraciones (en años):

#### **<font color="black">Calcular:</font>**

El estimador de máxima verosimilitud para $\theta$.


### **<font color="navy">Solución:</font>**

Sabemos que para una distribución exponencial:

X \sim \text{Exp}(\theta)
$$

el estimador de máxima verosimilitud es:

\hat{\theta}_{MLE} = \frac{1}{\bar{X}}

\bar{X} = \frac{1}{5}(2 + 3 + 1 + 3 + 5) = \frac{14}{5} = 2.8

## **<font color="purple">Ejercicio: </font>**
Calcular el estimador con Sympy.

```python
# LIBRERÍAS NECESARIAS

# sympy: para cálculo simbólico
import sympy as sp

# numpy: para manejo numérico de datos
import numpy as np


# Definimos el conjunto de datos observado
datos = np.array([2, 3, 1, 3, 5])

# Número de observaciones
n = len(datos)


# La media se calcula como: X̄ = (1/n) * suma de los datos
media = np.mean(datos)

print("Media muestral:", media)


# Definimos el parámetro θ como variable simbólica positiva
theta = sp.symbols('theta', positive=True)

# Creamos variables simbólicas x1, x2, ..., xn
x = sp.symbols(f'x1:{n+1}', positive=True)


# ℓ(θ) = n ln(θ) - θ * sum(x_i)
logL = n * sp.log(theta) - theta * sum(x)

print("\nLog-verosimilitud:")
sp.pprint(logL)


# Derivamos la log-verosimilitud respecto a θ
dlogL = sp.diff(logL, theta)

print("\nDerivada de la log-verosimilitud:")
sp.pprint(dlogL)


# Igualamos la derivada a cero y resolvemos
theta_hat = sp.solve(dlogL, theta)[0]

print("\nEstimador θ̂ :")
sp.pprint(theta_hat)


# Creamos un diccionario para sustituir x_i por los datos reales
subs_dict = {x[i]: datos[i] for i in range(n)}

# Evaluamos el estimador con los datos
theta_val = theta_hat.subs(subs_dict)

print("\nValor numérico de θ̂:" )
print(float(theta_val))


# Teóricamente: θ̂ = 1 / media
print("\nValidación (1/media):")
print(1 / media)
```

En particular, se han construido estimadores como:

$$
\hat{\theta}
$$

a partir de una muestra aleatoria:

X_1, X_2, \dots, X_n
$$

Una vez obtenido un estimador, surge una cuestión fundamental:

> **¿Cómo varía este estimador de una muestra a otra?**

Si se toman múltiples muestras de la misma población, cada una producirá un valor distinto de:

$$
\hat{\theta}
$$

Por lo tanto, el estimador no es un valor fijo, sino una **variable aleatoria**.


El comportamiento probabilístico de un estadístico se describe mediante su **distribución muestral**.


## **<font color="navy">Distribuciones Muestrales</font>**

En inferencia estadística, los parámetros poblacionales (como $\mu$ o $\sigma^2$) son desconocidos, por lo que se utilizan muestras para estimarlos.

A partir de una muestra:

$$
X_1, X_2, \dots, X_n
$$

se construyen estadísticos, como la media muestral:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$


Aunque al calcular $\bar{X}$ obtenemos un número, en realidad:

> $\bar{X}$ es una **variable aleatoria**

porque depende de la muestra seleccionada.


### **<font color="darkgreen">¿Por qué es importante?</font>**

Si tomamos muchas muestras distintas de la misma población:

- Cada muestra produce un valor diferente de $\bar{X}$  
- Estos valores forman una distribución  

Esa distribución se llama:

> **Distribución muestral**

La **<font color="crimson">distribución muestral</font>** de un estadístico es la distribución de probabilidad de ese estadístico cuando se consideran todas las posibles muestras de tamaño $n$.

El caso más importante es el de la media muestral:

$$
\bar{X} = \frac{1}{n} \sum_{i=1}^{n} X_i
$$

#### **<font color="brown">Propiedades</font>**

Si:

$$
X_i \sim (\mu, \sigma^2)
$$

entonces:

#### **Media:**

$$
\mathbb{E}[\bar{X}] = \mu
$$

La media muestral es un estimador **insesgado**


#### **Varianza:**

$$
\mathbb{V}[\bar{X}] = \frac{\sigma^2}{n}
$$

A mayor tamaño de muestra, menor variabilidad


- La media muestral fluctúa alrededor de $\mu$  
- Mientras más grande sea $n$, más estable será  


## **<font color="navy">Distribución Chi-Cuadrada</font>**

La **<font color="crimson">Distribución Chi-cuadrada</font>** es una de las distribuciones muestrales más importantes en inferencia estadística, especialmente cuando se trabaja con **varianza**.


### **<font color="darkgreen">¿De dónde sale la Chi-cuadrada?</font>**

Para entenderla, partimos de algo que ya conoces:

$$
Z \sim N(0,1)
$$

Si elevamos al cuadrado:

$$
Z^2
$$

obtenemos una variable que:

- siempre es positiva  
- mide “qué tan lejos está Z del 0” (en términos cuadrados)  

En general, si tenemos:

$$
Z_1, Z_2, \dots, Z_n \sim N(0,1)
$$

independientes, y hacemos:

$$
X = Z_1^2 + Z_2^2 + \cdots + Z_n^2
$$

entonces:

$$
X \sim \chi^2(n)
$$


La Chi-cuadrada es:

> la suma de “distancias cuadradas” de variables normales


### **<font color="darkgreen">Grados de libertad</font>**

El parámetro $n$ se llama:

> **grados de libertad**

Es el número de variables independientes que estás sumando.

Ejemplo:

- 1 normal → $\chi^2(1)$  
- 5 normales → $\chi^2(5)$  


### **<font color="darkgreen">Relación con la Varianza</font>**

Aquí es donde la Chi-cuadrada se vuelve importante.

Si:

$$
X_1, X_2, \dots, X_n \sim N(\mu, \sigma^2)
$$

entonces la varianza muestral:

$$
S^2 = \frac{1}{n-1} \sum (X_i - \bar{X})^2
$$

cumple:

$$
\frac{(n-1)S^2}{\sigma^2} \sim \chi^2(n-1)
$$


Esto significa que:

- La varianza muestral está basada en **cuadrados de desviaciones**  
- Esos cuadrados generan una distribución Chi-cuadrada  


### **<font color="darkgreen">¿Por qué aparece (n-1)?</font>**

Porque usamos la media muestral $\bar{X}$, no $\mu$.

Esto “consume” un grado de libertad


### **<font color="darkgreen">Forma de la distribución</font>**

La Chi-cuadrada tiene características muy importantes:



#### **<font color="purple">1. Siempre positiva</font>**

$$
X \geq 0
$$

#### **<font color="purple">2. Asimétrica</font>**

- Sesgada a la derecha  
- Cola larga hacia valores grandes  

#### **<font color="purple">3. Depende de n</font>**

- n pequeño → muy sesgada  
- n grande → se parece a una normal  


### **<font color="darkgreen">Propiedades</font>**

Si:

$$
X \sim \chi^2(n)
$$

entonces:


### **Media:**

$$
\mathbb{E}[X] = n
$$


### **Varianza:**

$$
\mathbb{V}[X] = 2n
$$

La Chi-cuadrada mide:

> cuánta variabilidad total hay en los datos



### **<font color="darkgreen">¿Para qué sirve?</font>**

Se utiliza principalmente para:

- Intervalos de confianza para la varianza
$
\sigma^2
$

- Pruebas de hipótesis sobre varianza

- Bondad de ajuste: ver si los datos siguen una distribución

```python
# IMPORTACIÓN DE LIBRERÍAS

import numpy as np                      # Para generar valores en x
import matplotlib.pyplot as plt         # Para graficar
from scipy.stats import chi2            # Distribución Chi-cuadrada


# EJE X
x = np.linspace(0, 20, 100)

# DIFERENTES VALORES DE n (grados de libertad)
n_values = [1, 3, 6, 10, 30]

# CREACIÓN DE LA GRÁFICA
plt.figure(figsize=(10,5))

# GRAFICAR CADA DISTRIBUCIÓN
for n in n_values:

    # PDF de la chi-cuadrada
    y = chi2.pdf(x, df=n)

    # Graficar curva
    plt.plot(x, y, label=f"Chi-cuadrada (n={n})")


# TÍTULO Y ETIQUETAS
plt.title("Distribución Chi-cuadrada")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")

# LEYENDA
plt.legend()

# MOSTRAR GRÁFICA
plt.show()
```

## **<font color="navy">Distribución t de Student</font>**

La **<font color="crimson">Distribución t de Student</font>** es una de las distribuciones muestrales más importantes en inferencia estadística, especialmente cuando se trabaja con la **media** y la **varianza es desconocida**.


### **<font color="darkgreen">¿De dónde sale la t de Student?</font>**

Para entenderla, partimos de dos resultados conocidos:

$$
Z \sim N(0,1)
$$

y

$$
V \sim \chi^2(n)
$$

independientes.

Si combinamos ambas variables:

$$
T = \frac{Z}{\sqrt{V/n}}
$$

entonces:

$$
T \sim t(n)
$$


La distribución t surge al dividir una normal estándar entre la raíz de una Chi-cuadrada normalizada.


### **<font color="darkgreen">Forma más importante</font>**

Si:

$$
X_1, X_2, \dots, X_n \sim N(\mu, \sigma^2)
$$

entonces:

$$
T = \frac{\bar{X} - \mu}{S / \sqrt{n}} \sim t(n-1)
$$


Esto significa que:

- Se reemplaza $\sigma$ por la desviación muestral $S$  
- Se introduce más variabilidad  
- Se utiliza la distribución t en lugar de la normal  


### **<font color="darkgreen">Grados de libertad</font>**

El parámetro $n-1$ se llama:

> **grados de libertad**

porque se pierde un grado de libertad al estimar la media $\bar{X}$.



### **<font color="darkgreen">¿Por qué no usamos la normal?</font>**

Cuando $\sigma$ es desconocida:

- Hay incertidumbre adicional  
- La distribución de la media ya no es exactamente normal  

Por eso se usa la distribución t, que tiene:

> **colas más pesadas**


### **<font color="darkgreen">Forma de la distribución</font>**

La t de Student tiene características importantes:


#### **<font color="purple">1. Simétrica</font>**

- Centrada en 0  
- Igual a ambos lados  


#### **<font color="purple">2. Colas más pesadas</font>**

- Mayor probabilidad de valores extremos  
- Refleja incertidumbre adicional  


#### **<font color="purple">3. Depende de n</font>**

- n pequeño → colas muy pesadas  
- n grande → se aproxima a normal  


### **<font color="darkgreen">Propiedades</font>**

Si:

$$
T \sim t(n)
$$

entonces:


### **Media:**

$$
\mathbb{E}[T] = 0
$$


### **Varianza:**

$$
\mathbb{V}[T] = \frac{n}{n-2}, \quad n > 2
$$


La t de Student mide:

> la variabilidad de la media cuando la varianza es desconocida


### **<font color="darkgreen">¿Para qué sirve?</font>**

Se utiliza principalmente para:

- Intervalos de confianza para la media
$
\mu
$

- Pruebas de hipótesis sobre la media

- Análisis con muestras pequeñas


### **<font color="darkgreen">Idea clave</font>**

Cuando no conoces la varianza poblacional:

> usas t de Student en lugar de la normal

```python
# IMPORTACIÓN DE LIBRERÍAS

import numpy as np                      # Para generar valores en x
import matplotlib.pyplot as plt         # Para graficar
from scipy.stats import t, norm         # Distribuciones t y normal


# EJE X
x = np.linspace(-5, 5, 1000)

# DIFERENTES VALORES DE n (grados de libertad)
n_values = [1, 3, 20, 10, 30]

# CREACIÓN DE LA GRÁFICA
plt.figure(figsize=(10,5))

# GRAFICAR DISTRIBUCIONES t
for n in n_values:

    y = t.pdf(x, df=n)

    plt.plot(x, y, label=f"t (gl={n})")


# GRAFICAR NORMAL ESTÁNDAR
y_norm = norm.pdf(x)

plt.plot(x, y_norm, linestyle='--', color='black', label="Normal estándar")


# TÍTULO Y ETIQUETAS
plt.title("Distribución t de Student")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")

# LEYENDA
plt.legend()

# MOSTRAR GRÁFICA
plt.show()
```

## **<font color="navy">Distribución F de Fisher</font>**

La **<font color="crimson">Distribución F de Fisher</font>** es una distribución muestral fundamental en inferencia estadística, especialmente cuando se comparan **varianzas**.

### **<font color="darkgreen">¿De dónde sale la distribución F?</font>**

Partimos de dos variables independientes con distribución Chi-cuadrada:

$$
X_1 \sim \chi^2(n_1), \quad X_2 \sim \chi^2(n_2)
$$

Si construimos el cociente:

$$
F = \frac{X_1 / n_1}{X_2 / n_2}
$$

entonces:

$$
F \sim F(n_1, n_2)
$$


La distribución F es:

> el cociente de dos variables Chi-cuadrada normalizadas


### **<font color="darkgreen">Grados de libertad</font>**

La distribución F tiene **dos parámetros**:

- $n_1$: grados de libertad del numerador  
- $n_2$: grados de libertad del denominador  



### **<font color="darkgreen">Relación con la Varianza</font>**

Si:

$$
X_1, \dots, X_{n_1} \sim N(\mu_1, \sigma_1^2)
$$

$$
X_2, \dots, X_{n_2} \sim N(\mu_2, \sigma_2^2)
$$

entonces:

$$
\frac{S_1^2 / \sigma_1^2}{S_2^2 / \sigma_2^2} \sim F(n_1 - 1, n_2 - 1)
$$


Esto significa que:

- La distribución F compara dos varianzas  
- Es útil para evaluar si dos poblaciones tienen la misma variabilidad  


### **<font color="darkgreen">Forma de la distribución</font>**

La distribución F tiene características importantes:


#### **<font color="purple">1. Siempre positiva</font>**

$$
F \geq 0
$$


#### **<font color="purple">2. Asimétrica</font>**

- Sesgada a la derecha  
- Cola larga hacia valores grandes  


#### **<font color="purple">3. Depende de $n_1$ y $n_2$</font>**

- Valores pequeños → más dispersa  
- Valores grandes → más concentrada  


### **<font color="darkgreen">Propiedades</font>**

Si:

$$
F \sim F(n_1, n_2)
$$

entonces:


### **Media:**

$$
\mathbb{E}[F] = \frac{n_2}{n_2 - 2}, \quad n_2 > 2
$$


### **Varianza:**

$$
\mathbb{V}[F] = \frac{2 n_2^2 (n_1 + n_2 - 2)}{n_1 (n_2 - 2)^2 (n_2 - 4)}, \quad n_2 > 4
$$

### **<font color="darkgreen">¿Para qué sirve?</font>**

Se utiliza principalmente para:

- Comparar varianzas  
- Análisis de varianza (ANOVA)  
- Pruebas de hipótesis  

```python
# IMPORTACIÓN DE LIBRERÍAS

import numpy as np                      # Para generar valores en x
import matplotlib.pyplot as plt         # Para graficar
from scipy.stats import f               # Distribución F de Fisher


# EJE X
x = np.linspace(-4, 5, 100)

# DIFERENTES PARES DE GRADOS DE LIBERTAD (n1, n2)
parametros = [(1,2), (2,5), (5,10), (10,20)]

# CREACIÓN DE LA GRÁFICA
plt.figure(figsize=(10,5))

# GRAFICAR CADA DISTRIBUCIÓN
for n1, n2 in parametros:

    y = f.pdf(x, dfn=n1, dfd=n2)

    plt.plot(x, y, label=f"F (gl1={n1}, gl2={n2})")


# TÍTULO Y ETIQUETAS
plt.title("Distribución F de Fisher")
plt.xlabel("x")
plt.ylabel("Densidad de probabilidad")

# LEYENDA
plt.legend()

# MOSTRAR GRÁFICA
plt.show()
```

## **<font color="navy">Intervalos de Confianza</font>**

En inferencia estadística, un estimador puntual como $\bar{X}$ proporciona un solo valor para aproximar un parámetro poblacional.

Sin embargo, debido a la variabilidad de las muestras, ese valor cambia de una muestra a otra.

Por esta razón, es más informativo construir un **intervalo de valores plausibles** para el parámetro.

Un intervalo de confianza responde a la pregunta:

> ¿En qué rango es razonable que se encuentre el parámetro poblacional?

### **<font color="darkgreen">Definición</font>**

Un **<font color="crimson">intervalo de confianza</font>** es un **intervalo aleatorio** construido a partir de una muestra, cuyos extremos dependen de los datos observados:

$$
(L(X_1,\dots,X_n), \; U(X_1,\dots,X_n))
$$

tal que cumple:

$$
P\big(L \leq \theta \leq U\big) = 1 - \alpha
$$

El parámetro $\theta$ es **fijo pero desconocido**, mientras que el intervalo es **aleatorio** porque depende de la muestra.

Por lo tanto:

> La probabilidad $1 - \alpha$ se refiere al **procedimiento de construcción del intervalo**, no al parámetro en sí.

Un intervalo de confianza del 95% NO significa:

“El parámetro tiene 95% de probabilidad de estar en el intervalo”

Significa:

> Si repetimos el experimento muchas veces, el 95% de los intervalos construidos contendrán al verdadero parámetro.

### **<font color="darkgreen">Construcción del Intervalo de Confianza</font>**

La construcción de un intervalo de confianza se basa en una idea fundamental:

> transformar un problema desconocido en uno conocido usando una variable aleatoria con distribución conocida

Sabemos que:

$$
Z = \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \sim N(0,1)
$$

- $\bar{X}$ es aleatoria  
- $\mu$ es desconocido pero fijo  
- $Z$ tiene distribución conocida  

Para la normal estándar sabemos que:

$$
P(-z_{\alpha/2} \leq Z \leq z_{\alpha/2}) = 1 - \alpha
$$

Esto significa que:

- El área central bajo la curva es $1 - \alpha$  
- Las colas tienen probabilidad $\alpha/2$ cada una  

Sustituimos la expresión de $Z$:

$$
P\left(
- z_{\alpha/2} \leq \frac{\bar{X} - \mu}{\sigma / \sqrt{n}} \leq z_{\alpha/2}
\right) = 1 - \alpha
$$

Estamos expresando una probabilidad sobre $\mu$ usando algo que sí conocemos.

Multiplicamos todo por $\sigma / \sqrt{n}$:

$$
P\left(
- z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \leq \bar{X} - \mu \leq z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
\right) = 1 - \alpha
$$

Sumamos $\mu$ y restamos los términos:

$$
P\left(
\bar{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}} \leq \mu \leq
\bar{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
\right) = 1 - \alpha
$$

**Se tiene como resultado:**

$$
\mu \in
\left(
\bar{X} - z_{\alpha/2} \frac{\sigma}{\sqrt{n}},
\;
\bar{X} + z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
\right)
$$

- El intervalo depende de $\bar{X}$ → es aleatorio  
- $\mu$ es fijo → lo queremos capturar  

La probabilidad está en el intervalo, no en $\mu$


#### **<font color="darkgreen">En resumen: </font>**

1. Partes de una variable con distribución conocida  
2. Usas una probabilidad central  
3. Sustituyes el estadístico  
4. Despejas el parámetro  
5. Obtienes el intervalo  

### **<font color="darkgreen">Estructura General del Intervalo de Confianza</font>**

Todo intervalo de confianza se construye bajo la misma forma:

$$
\text{Estimador} \;\pm\; \text{Margen de error}
$$

La idea es muy intuitiva:

> Tomamos el mejor valor estimado (estimador) y le agregamos un “colchón de incertidumbre” (error)


#### **<font color="purple">1. Estimador</font>**

Es el valor central del intervalo.

Ejemplo:

$$
\bar{X}
$$

- Es el mejor estimado del parámetro  
- Es el centro del intervalo  
- Depende de los datos → es aleatorio  

#### **<font color="purple">2. Margen de error</font>**

Se define como:

$$
\text{Error} = \text{Valor critico} \times \text{Error estandar}
$$


##### **<font color="purple">2.1. Error estándar</font>**

$$
\text{EE} = \frac{\sigma}{\sqrt{n}} \quad \text{o} \quad \frac{S}{\sqrt{n}}
$$


- Mide la variabilidad del estimador  
- Indica qué tanto cambia $\bar{X}$ entre muestras  
- Disminuye cuando aumenta $n$  

##### **<font color="purple">2.2 Valor crítico</font>**

Depende de la distribución y del nivel de confianza:


#### **Normal:**

$$
z_{\alpha/2}
$$

#### **t de Student:**

$$
t_{\alpha/2, n-1}
$$

#### **Chi-cuadrada:**

Usa cuantiles asimétricos:

$$
\chi^2_{\alpha/2}, \quad \chi^2_{1-\alpha/2}
$$

- Define qué tan “ancho” es el intervalo  
- A mayor confianza → mayor valor crítico  


### **<font color="darkgreen">¿Qué controla el ancho del intervalo?</font>**

##### **Tamaño de muestra**

$$
n \uparrow \Rightarrow \text{EE} \downarrow \Rightarrow \text{intervalo mas pequeño}
$$

##### **Nivel de confianza**

$$
1-\alpha \uparrow \Rightarrow \text{valor crítico} \uparrow \Rightarrow \text{intervalo mas grande}
$$

##### **Variabilidad**

$$
\sigma \uparrow \Rightarrow \text{intervalo más amplio}
$$

**Caso especial: Chi-cuadrada**

Aquí la estructura cambia ligeramente:

- No es simétrica  
- Se construye con dos cuantiles distintos  


## **<font color="navy">Casos de Intervalos de Confianza</font>**

### **<font color="darkgreen">a) Intervalo para la media $\mu$ ( $\sigma$ conocida)</font>**

$$
\mu \in
\left(
\bar{X} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
\right)
$$

Se estima el valor de la media poblacional cuando la variabilidad poblacional es conocida.

El intervalo es simétrico alrededor de $\bar{X}$ y su amplitud depende únicamente de $n$ y del nivel de confianza.

### **<font color="darkgreen">b) Intervalo para la media $\mu$ ($\sigma$ desconocida)</font>**

#### **Caso 1: Muestras grandes ($n > 30$)**

$$
\mu \in
\left(
\bar{X} \pm z_{\alpha/2} \frac{S}{\sqrt{n}}
\right)
$$

#### **Caso 2: Muestras pequeñas ($n \leq 30$)**

$$
\mu \in
\left(
\bar{X} \pm t_{\alpha/2,\,n-1} \frac{S}{\sqrt{n}}
\right)
$$

Se estima la media poblacional cuando la varianza es desconocida.

- Para muestras grandes, se usa la normal por el Teorema Central del Límite  
- Para muestras pequeñas, se usa la t de Student debido a la mayor incertidumbre  

El intervalo es simétrico alrededor de $\bar{X}$ pero más ancho en el caso t.


### **<font color="darkgreen">c) Intervalo para la varianza $\sigma^2$</font>**

$$
\sigma^2 \in
\left(
\frac{(n-1)S^2}{\chi^2_{1-\alpha/2,\,n-1}},
\;
\frac{(n-1)S^2}{\chi^2_{\alpha/2,\,n-1}}
\right)
$$


Se estima la dispersión poblacional.

El intervalo no es simétrico, ya que depende de la distribución Chi-cuadrada.

Su forma refleja la asimetría de la distribución de la varianza.


### **<font color="darkgreen">d) Intervalo para la diferencia de medias ($\mu_1 - \mu_2$)</font>**

#### **Varianzas conocidas**

$$
\mu_1 - \mu_2 \in
\left(
(\bar{X} - \bar{Y}) \pm z_{\alpha/2}
\sqrt{\frac{\sigma_1^2}{n_1} + \frac{\sigma_2^2}{n_2}}
\right)
$$


#### **Varianzas desconocidas**

##### **Caso 1: Muestras grandes**

$$
\mu_1 - \mu_2 \in
\left(
(\bar{X} - \bar{Y}) \pm z_{\alpha/2}
\sqrt{\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}}
\right)
$$

##### **Caso 2: Varianzas iguales (pooled)**

$$
\mu_1 - \mu_2 \in
\left(
(\bar{X} - \bar{Y}) \pm t_{\alpha/2,\,n_1+n_2-2}
\cdot S_p \sqrt{\frac{1}{n_1} + \frac{1}{n_2}}
\right)
$$

###### **Definición de $S_p$**

$$
S_p^2 =
\frac{(n_1 - 1)S_1^2 + (n_2 - 1)S_2^2}{n_1 + n_2 - 2}
$$

- Es la **varianza combinada (pooled)**  
- Promedia las varianzas muestrales ponderadas  
- Se usa cuando se asume que $\sigma_1^2 = \sigma_2^2$  

##### **Caso 3: Varianzas distintas (Welch)**

$$
\mu_1 - \mu_2 \in
\left(
(\bar{X} - \bar{Y}) \pm t_{\alpha/2,\,f}
\sqrt{\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}}
\right)
$$
###### **Grados de libertad aproximados (Welch)**

$$
f =
\frac{\left(\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}\right)^2}
{\frac{\left(\frac{S_1^2}{n_1}\right)^2}{n_1 - 1} +
 \frac{\left(\frac{S_2^2}{n_2}\right)^2}{n_2 - 1}}
$$

Se evalúa la diferencia entre dos medias poblacionales.

El intervalo es simétrico alrededor de $(\bar{X} - \bar{Y})$.

La presencia o no del valor 0 determina si existe evidencia de diferencia entre las medias.


### **<font color="darkgreen">e) Intervalo para la razón de varianzas</font>**

$$
\frac{\sigma_1^2}{\sigma_2^2} \in
\left(
\frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{1-\alpha/2}},
\;
\frac{S_1^2}{S_2^2} \cdot \frac{1}{F_{\alpha/2}}
\right)
$$


Se compara la variabilidad relativa entre dos poblaciones.

El intervalo es asimétrico.

El valor 1 es el punto de referencia para igualdad de varianzas.


### **<font color="darkgreen">f) Intervalo para proporción $p$</font>**

$$
p \in
\left(
\hat{p} \pm z_{\alpha/2}
\sqrt{\frac{\hat{p}(1 - \hat{p})}{n}}
\right)
$$

Se estima una probabilidad poblacional.

El intervalo es simétrico alrededor de $\hat{p}$.

Su amplitud depende de la variabilidad binomial y del tamaño de muestra.


### **<font color="darkgreen">g) Intervalo para diferencia de proporciones</font>**

$$
p_1 - p_2 \in
\left(
(\hat{p}_1 - \hat{p}_2) \pm z_{\alpha/2}
\sqrt{
\frac{\hat{p}_1(1 - \hat{p}_1)}{n_1} +
\frac{\hat{p}_2(1 - \hat{p}_2)}{n_2}
}
\right)
$$

Se compara la diferencia entre dos proporciones poblacionales.

El intervalo es simétrico alrededor de $(\hat{p}_1 - \hat{p}_2)$.

El valor 0 indica ausencia de diferencia entre proporciones.


## **<font color="purple">Ejemplo: Intervalo de Confianza para la Media ($\mu$) con $\sigma$ conocida</font>**


Una compañía de seguros está interesada en estimar el **tiempo promedio de resolución de siniestros** (en días) para cierto tipo de póliza.

A partir de registros históricos, se sabe que la desviación estándar poblacional es:

$$
\sigma = 12 \text{ días}
$$

Se selecciona una muestra aleatoria de:

$$
n = 36
$$

siniestros recientes, obteniendo una media muestral de:

$$
\bar{X} = 50 \text{ días}
$$

Se desea construir un intervalo de confianza del **95%** para la media poblacional $\mu$.

### **<font color="navy">Solución</font>**

**Como la desviación estándar poblacional es conocida, se utiliza:**

$$
Z \sim N(0,1)
$$

**Nivel de confianza:**

$$
1 - \alpha = 0.95 \Rightarrow \alpha = 0.05
$$

**Valor crítico:**

$$
z_{\alpha/2} = z_{0.025} = 1.96
$$

**Error estándar:**

$$
\frac{\sigma}{\sqrt{n}} = \frac{12}{\sqrt{36}} = 2
$$

**Margen de error:**

$$
E = 1.96 \cdot 2 = 3.92
$$

**Intervalo de confianza:**

$$
\mu \in (50 - 3.92,\; 50 + 3.92)
$$

$$
\mu \in (46.08,\; 53.92)
$$


> Con un nivel de confianza del 95%, el tiempo promedio de resolución de siniestros se encuentra entre:

$$
46.08 \text{ y } 53.92 \text{ días}
$$

```python
# INTERVALO DE CONFIANZA PARA LA MEDIA (σ CONOCIDA)

# importamos librerías necesarias
import numpy as np
from scipy.stats import norm


# DATOS
# n: tamaño de la muestra
n = 36

# x_bar: media muestral
x_bar = 50

# sigma: desviación estándar conocida
sigma = 12

# conf: nivel de confianza
conf = 0.95


# CÁLCULOS
# alpha: nivel de significancia
alpha = 1 - conf

# valor crítico Z de la normal estándar
z = norm.ppf(1 - alpha/2)

# error estándar de la media
se = sigma / np.sqrt(n)

# margen de error
E = z * se

# límites del intervalo
LI = x_bar - E
LS = x_bar + E


# RESULTADOS
# imprimimos el error estándar
print(f"Error estándar: {se:.2f}")

# imprimimos el margen de error
print(f"Margen de error: {E:.2f}\n")

# imprimimos el intervalo final
print(f"El intervalo de confianza al {int(conf*100)}% para la media es: ({LI:.2f}, {LS:.2f})")
```

## **<font color="purple">Ejemplo: Intervalo de Confianza para la Media ($\mu$) con $\sigma$ desconocida</font>**


Una compañía de seguros desea estimar el **monto promedio de los siniestros** (en miles de pesos) para una determinada cartera.

Se toma una muestra aleatoria de:

$$
n = 25
$$

siniestros, obteniendo una media:

$$
\bar{X} = 50
$$

y una desviación estándar muestral:

$$
S = 12
$$

Construir un intervalo de confianza del **95%** para la media poblacional $\mu$.

### **<font color="navy">Solución</font>**

**Como la varianza poblacional es desconocida, se utiliza la distribución t de Student:**

$$
T \sim t(n-1)
$$

**Nivel de confianza:**

$$
1 - \alpha = 0.95 \Rightarrow \alpha = 0.05
$$

**Grados de libertad:**

$$
df = n - 1 = 25 - 1 = 24
$$

**Valor crítico:**

De la distribución t:

$$
t_{\alpha/2,\,24} = t_{0.025,\,24} \approx 2.064
$$

**Error estándar:**

$$
\frac{S}{\sqrt{n}} = \frac{12}{\sqrt{25}} = \frac{12}{5} = 2.4
$$

**Margen de error:**

$$
E = t_{\alpha/2,\,24} \cdot \frac{S}{\sqrt{n}} = 2.064 \cdot 2.4 = 4.95
$$

**Intervalo de confianza:**

$$
\mu \in
\left(
50 - 4.95,\; 50 + 4.95
\right)
$$

$$
\mu \in (45.05,\; 54.95)
$$

> Con un nivel de confianza del 95%, la media poblacional $\mu$ se encuentra dentro del intervalo:

$$
(45.05,\; 54.95)
$$

```python
# INTERVALO DE CONFIANZA PARA LA MEDIA (σ DESCONOCIDA)

# importamos librerías necesarias
import numpy as np
from scipy.stats import t


# DATOS
# n: tamaño de la muestra
n = 25

# x_bar: media muestral
x_bar = 50

# s: desviación estándar muestral
s = 12

# conf: nivel de confianza
conf = 0.95


# VALIDACIONES
# verificar que n sea válido
if n <= 1:
    raise ValueError("n debe ser mayor que 1")

# verificar que s sea positiva
if s <= 0:
    raise ValueError("s debe ser mayor que 0")

# verificar nivel de confianza válido
if conf <= 0 or conf >= 1:
    raise ValueError("conf debe estar entre 0 y 1")


# CÁLCULOS
# alpha: nivel de significancia
alpha = 1 - conf

# grados de libertad
df = n - 1

# valor crítico t de Student
t_crit = t.ppf(1 - alpha/2, df)

# error estándar de la media
se = s / np.sqrt(n)

# margen de error
E = t_crit * se

# límites del intervalo
LI = x_bar - E
LS = x_bar + E


# RESULTADOS
# imprimimos el error estándar
print(f"Error estándar: {se:.2f}")

# imprimimos el margen de error
print(f"Margen de error: {E:.2f}\n")

# imprimimos el intervalo final
print(f"El intervalo de confianza al {int(conf*100)}% para la media es:")
print(f"({LI:.2f}, {LS:.2f})")
```

### **<font color="darkgreen">Uso de la función `interval` en Python</font>**

En Python, la función `interval` de `scipy.stats` permite calcular intervalos de confianza de forma directa, sin necesidad de obtener manualmente el valor crítico ni construir la fórmula paso a paso.

La sintaxis general es:

$$
\text{intervalo} = \text{dist.interval}(\text{conf}, \; \text{parámetros})
$$

### **Interpretación de los argumentos**

- `conf`: nivel de confianza  
- `loc`: estimador (por ejemplo, $\bar{X}$)  
- `scale`: error estándar  
- `df`: grados de libertad (en distribuciones como t)


La función `interval` automatiza la expresión:

$$
\text{Estimador} \pm \text{Valor crítico} \times \text{Error estándar}
$$

```python
# INTERVALO DE CONFIANZA PARA LA MEDIA (σ DESCONOCIDA) CON interval()

# importamos librerías necesarias
import numpy as np
from scipy.stats import t


# DATOS
# n: tamaño de la muestra
n = 25

# x_bar: media muestral
x_bar = 50

# s: desviación estándar muestral
s = 12

# conf: nivel de confianza
conf = 0.95


# CÁLCULOS
# grados de libertad
df = n - 1

# error estándar
se = s / np.sqrt(n)

# intervalo usando t
LI, LS = t.interval(conf, df, loc=x_bar, scale=se)


# RESULTADOS
# imprimimos el error estándar
print(f"Error estándar: {se:.2f}\n")

# imprimimos el intervalo
print(f"El intervalo de confianza al {int(conf*100)}% para la media es:")
print(f"({LI:.2f}, {LS:.2f})")
```

## **<font color="purple">Ejemplo: Intervalo de Confianza para la Varianza ($\sigma^2$)</font>**

Una compañía de seguros desea analizar la **variabilidad en los montos de los siniestros** para una determinada cartera.

Se toma una muestra aleatoria de:

$$
n = 20
$$

siniestros, obteniendo una varianza muestral de:

$$
S^2 = 144
$$

Construir un intervalo de confianza del **95%** para la varianza poblacional $\sigma^2$.

### **<font color="navy">Solución:</font>**

**Para la varianza, se utiliza la distribución Chi-cuadrada:**

$$
\frac{(n-1)S^2}{\sigma^2} \sim \chi^2(n-1)
$$

**Nivel de confianza:**

$$
1 - \alpha = 0.95 \Rightarrow \alpha = 0.05
$$

**Grados de libertad:**

$$
df = n - 1 = 20 - 1 = 19
$$

**Valores críticos:**

De la distribución Chi-cuadrada:

$$
\chi^2_{0.025,\,19} \approx 32.85
$$

$$
\chi^2_{0.975,\,19} \approx 8.91
$$

Construyendo el intervalo:

$$
\sigma^2 \in
\left(
\frac{(n-1)S^2}{\chi^2_{0.025}},
\;
\frac{(n-1)S^2}{\chi^2_{0.975}}
\right)
$$

Sustituyendo:

$$
\sigma^2 \in
\left(
\frac{19 \cdot 144}{32.85},
\;
\frac{19 \cdot 144}{8.91}
\right)
$$

**El intervalo de confianza es:**

$$
\sigma^2 \in (83.3,\; 307.0)
$$


> Con un nivel de confianza del 95%, la varianza poblacional se encuentra en el intervalo:

$$
(83.3,\; 307.0)
$$

```python
# INTERVALO DE CONFIANZA PARA LA VARIANZA (σ²) CON χ²

# importamos librerías necesarias
import numpy as np
from scipy.stats import chi2


# DATOS
# n: tamaño de la muestra
n = 20

# s2: varianza muestral
s2 = 144

# conf: nivel de confianza
conf = 0.95


# VALIDACIONES
# verificar que n sea válido
if n <= 1:
    raise ValueError("n debe ser mayor que 1")

# verificar que la varianza sea positiva
if s2 <= 0:
    raise ValueError("s2 debe ser mayor que 0")

# verificar nivel de confianza válido
if conf <= 0 or conf >= 1:
    raise ValueError("conf debe estar entre 0 y 1")


# CÁLCULOS
# alpha: nivel de significancia
alpha = 1 - conf

# grados de libertad
df = n - 1

# valores críticos de chi-cuadrada
chi2_inf = chi2.ppf(alpha/2, df)
chi2_sup = chi2.ppf(1 - alpha/2, df)

# construcción del intervalo
LI = (df * s2) / chi2_sup
LS = (df * s2) / chi2_inf


# RESULTADOS
# imprimimos el intervalo final
print(f"El intervalo de confianza al {int(conf*100)}% para la varianza es:")
print(f"({LI:.2f}, {LS:.2f})")
```

## **<font color="purple">Ejemplo: Intervalo de Confianza para la Diferencia de Medias ($\mu_1 - \mu_2$)</font>**

Una compañía de seguros desea comparar el **monto promedio de siniestros** entre dos tipos de pólizas.

Se toman dos muestras independientes:

Grupo 1:
$$
n_1 = 30, \quad \bar{X}_1 = 55, \quad S_1 = 10
$$

Grupo 2:
$$
n_2 = 25, \quad \bar{X}_2 = 50, \quad S_2 = 12
$$

Construir un intervalo de confianza del **95%** para la diferencia de medias:

$$
\mu_1 - \mu_2
$$


### **<font color="navy">Solución:</font>**

- Muestras independientes  
- Varianzas desconocidas  
- No se asume igualdad de varianzas  


**Se utiliza la aproximación de Welch:**

$$
T \sim t(f)
$$

**Diferencia de medias:**

$$
\bar{X}_1 - \bar{X}_2 = 55 - 50 = 5
$$

**Error estándar:**

$$
\sqrt{\frac{S_1^2}{n_1} + \frac{S_2^2}{n_2}} =
\sqrt{\frac{100}{30} + \frac{144}{25}} =
\sqrt{3.33 + 5.76} = \sqrt{9.09} \approx 3.01
$$

**Grados de libertad:**

$$
f =
\frac{(3.33 + 5.76)^2}
{\frac{(3.33)^2}{29} + \frac{(5.76)^2}{24}}
\approx 46
$$

**Valor crítico:**

$$
t_{0.025,\,46} \approx 2.013
$$

**Margen de error:**

$$
E = 2.013 \cdot 3.01 \approx 6.06
$$

**Intervalo de confianza:**

$$
\mu_1 - \mu_2 \in (5 - 6.06,\; 5 + 6.06)
$$

**El Intervalo de confianza es:**

$$
\mu_1 - \mu_2 \in (-1.06,\; 11.06)
$$

> El intervalo contiene el valor 0, lo que indica que no hay evidencia suficiente para afirmar que existe una diferencia significativa entre las medias poblacionales.


```python
# INTERVALO DE CONFIANZA PARA LA DIFERENCIA DE MEDIAS (WELCH)

# importamos librerías necesarias
import numpy as np
from scipy.stats import t


# DATOS
# grupo 1
n1 = 30
x1_bar = 55
s1 = 10

# grupo 2
n2 = 25
x2_bar = 50
s2 = 12

# nivel de confianza
conf = 0.95


# VALIDACIONES
# tamaños de muestra válidos
if n1 <= 1 or n2 <= 1:
    raise ValueError("n1 y n2 deben ser mayores que 1")

# desviaciones estándar positivas
if s1 <= 0 or s2 <= 0:
    raise ValueError("s1 y s2 deben ser mayores que 0")

# nivel de confianza válido
if conf <= 0 or conf >= 1:
    raise ValueError("conf debe estar entre 0 y 1")


# CÁLCULOS
# alpha: nivel de significancia
alpha = 1 - conf

# diferencia de medias
diff = x1_bar - x2_bar

# error estándar (Welch)
se = np.sqrt((s1**2)/n1 + (s2**2)/n2)

# grados de libertad aproximados
num = ((s1**2)/n1 + (s2**2)/n2)**2
den = ((s1**2)/n1)**2 / (n1 - 1) + ((s2**2)/n2)**2 / (n2 - 1)
df = num / den

# valor crítico t
t_crit = t.ppf(1 - alpha/2, df)

# margen de error
E = t_crit * se

# límites del intervalo
LI = diff - E
LS = diff + E


# RESULTADOS
# imprimimos el error estándar
print(f"Error estándar: {se:.2f}")

# imprimimos el margen de error
print(f"Margen de error: {E:.2f}\n")

# imprimimos el intervalo
print(f"El intervalo de confianza al {int(conf*100)}% para la diferencia de medias es:")
print(f"({LI:.2f}, {LS:.2f})")
```

## **<font color="purple">Ejemplo: Intervalo de Confianza para una Proporción ($p$)</font>**

Los anunciantes en televisión pueden creer erróneamente que casi todas las personas que ven TV entienden la mayor parte de los anuncios que ven y escuchan. Un estudio de investigación reciente pidió a 2300 personas de más de 13 años de edad que vieran extractos de publicidad de televisión de 30 segundos de duración. De éstos, 1914 televidentes entendieron mal todo o parte del extracto que vieron. Encuentre un intervalo de confianza de 95% para la proporción de todos los telespectadores (de los cuales la muestra es representativa) que entenderán mal el total o parte de los extractos de televisión empleados en este estudio.

### **<font color="navy">Solución: </font>**

**Proporción muestral:**

$$
\hat{p} = \frac{1914}{2300} \approx 0.8322
$$

**Nivel de confianza:**

$$
\alpha = 0.05, \quad z_{\alpha/2} = 1.96
$$

**Error estándar:**

$$
\sqrt{\frac{\hat{p}(1 - \hat{p})}{n}} =
\sqrt{\frac{0.8322(1 - 0.8322)}{2300}}
=
\sqrt{\frac{0.1397}{2300}} =
\sqrt{0.0000607} \approx 0.00779
$$

**Margen de error:**

$$
E = 1.96 \cdot 0.00779 \approx 0.0153
$$

**Intervalo de confianza:**

$$
p \in (0.8322 - 0.0153,\; 0.8322 + 0.0153)
$$

**El intervalo de confianza es:**

$$
p \in (0.8169,\; 0.8475)
$$

> Con un nivel de confianza del 95%, la proporción de telespectadores que no entienden total o parcialmente los anuncios se encuentra entre:

$$
81.69\% \text{ y } 84.75\%
$$

```python
# INTERVALO DE CONFIANZA PARA UNA PROPORCIÓN (p)
# UTILIZANDO STATSMODELS

# importamos la función especializada
# proportion_confint calcula intervalos para proporciones
from statsmodels.stats.proportion import proportion_confint


# DATOS
# x: número de éxitos (personas que NO entendieron)
x = 1914

# n: tamaño de la muestra
n = 2300

# nivel de confianza
conf = 0.95


# VALIDACIONES
# verificar tamaño de muestra válido
if n <= 0:
    raise ValueError("n debe ser mayor que 0")

# verificar número de éxitos válido
if x < 0 or x > n:
    raise ValueError("x debe estar entre 0 y n")

# verificar nivel de confianza válido
if conf <= 0 or conf >= 1:
    raise ValueError("conf debe estar entre 0 y 1")


# CÁLCULO DEL INTERVALO
# method="normal" corresponde al método Wald (clásico)
# alpha = 1 - nivel de confianza
LI, LS = proportion_confint(
    count=x,
    nobs=n,
    alpha=1 - conf,
    method="normal"
)


# RESULTADOS
# imprimimos el intervalo de forma clara
print(f"El intervalo de confianza al {int(conf*100)}% para la proporción es:")
print(f"({LI:.4f}, {LS:.4f})")


# INTERPRETACIÓN AUTOMÁTICA
# mostramos la proporción estimada
p_hat = x / n
print(f"\nProporción estimada: {p_hat:.4f}")

# interpretación sencilla
print("\nInterpretación:")
print(f"Entre {LI*100:.2f}% y {LS*100:.2f}% de la población no entiende completamente los anuncios.")
```

## **<font color="purple">Ejemplo: Intervalo de Confianza para la Razón de Varianzas ($\sigma_1^2 / \sigma_2^2$)</font>**


Un fabricante farmacéutico compra materias primas de dos proveedores diferentes. El nivel medio de impurezas es aproximadamente el mismo para ambos proveedores, pero el fabricante está preocupado por la variabilidad en la cantidad de impurezas de entre un embarque y otro. Si el nivel de impurezas tiende a variar en forma excesiva de una fuente de abastecimiento, esto podría afectar la calidad del producto final. Para comparar la variación en el porcentaje de impurezas para los dos proveedores, el fabricante selecciona diez envíos de cada uno de ellos y mide el porcentaje de impurezas de cada envío. Las varianzas muestrales fueron $s_1^2 = .273$ y $s_2^2 = .094$ respectivamente. Forme un intervalo de confianza de 95% para la relación entre las varianzas poblacionales reales.


### **<font color="navy">Solución</font>**

Se desea estimar:

$$
\frac{\sigma_1^2}{\sigma_2^2}
$$

Como se trabaja con varianzas de dos poblaciones normales, **se utiliza la distribución F:**

$$
\frac{S_1^2 / \sigma_1^2}{S_2^2 / \sigma_2^2} \sim F(df_1, df_2)
$$

**Datos y grados de libertad:**

$$
n_1 = 10, \quad n_2 = 10
$$

$$
df_1 = n_1 - 1 = 9, \quad df_2 = n_2 - 1 = 9
$$

**Razón muestral:**

$$
\frac{s_1^2}{s_2^2} = \frac{0.273}{0.094} \approx 2.904
$$

Este valor es el estimador puntual de la razón de varianzas.

**Nivel de confianza:**

$$
1 - \alpha = 0.95 \Rightarrow \alpha = 0.05
$$

Se trabajará con:

$$
\frac{\alpha}{2} = 0.025
$$

**Valores críticos de la distribución F:**

Se requieren dos cuantiles:

$$
F_{0.025,\,9,9} \approx 4.03
$$

$$
F_{0.975,\,9,9} \approx 0.248
$$


El intervalo de confianza para la razón de varianzas es:

$$
\frac{\sigma_1^2}{\sigma_2^2} \in
\left(
\frac{s_1^2/s_2^2}{F_{0.025}},
\;
\frac{s_1^2/s_2^2}{F_{0.975}}
\right)
$$

Sustituyendo:

$$
\left(
\frac{2.904}{4.03},\;
\frac{2.904}{0.248}
\right)
$$

**El intervalo de confianza es:**

$$
\frac{\sigma_1^2}{\sigma_2^2} \in (0.72,\; 11.71)
$$

El intervalo incluye el valor:

$$
1
$$

> Esto indica que no hay evidencia suficiente para afirmar que las varianzas poblacionales son diferentes.


```python
# INTERVALO DE CONFIANZA PARA LA RAZÓN DE VARIANZAS (σ1² / σ2²)

# importamos librerías necesarias
from scipy.stats import f


# DATOS
# varianzas muestrales
s1_2 = 0.273
s2_2 = 0.094

# tamaños de muestra
n1 = 10
n2 = 10

# nivel de confianza
conf = 0.95


# VALIDACIONES
# verificar tamaños de muestra válidos
if n1 <= 1 or n2 <= 1:
    raise ValueError("n1 y n2 deben ser mayores que 1")

# verificar varianzas positivas
if s1_2 <= 0 or s2_2 <= 0:
    raise ValueError("las varianzas deben ser mayores que 0")

# verificar nivel de confianza válido
if conf <= 0 or conf >= 1:
    raise ValueError("conf debe estar entre 0 y 1")


# CÁLCULOS
# grados de libertad
df1 = n1 - 1
df2 = n2 - 1

# razón muestral
ratio = s1_2 / s2_2

# intervalo de la distribución F
F_low, F_high = f.interval(conf, df1, df2)

# construcción del intervalo para σ1² / σ2²
LI = ratio / F_high
LS = ratio / F_low


# RESULTADOS
print(f"Razón muestral (s1^2 / s2^2): {ratio:.3f}\n")

print(f"El intervalo de confianza al {int(conf*100)}% para la razón de varianzas es:")
print(f"({LI:.2f}, {LS:.2f})")


# INTERPRETACIÓN
if LI <= 1 <= LS:
    print("\nEl intervalo contiene 1 → no hay diferencia significativa entre varianzas.")
else:
    print("\nEl intervalo NO contiene 1 → las varianzas son diferentes.")
```

```python
# INTERVALOS DE CONFIANZA (MEDIA Y VARIANZA) USANDO interval()

# importamos librerías necesarias
import numpy as np
from scipy.stats import norm, t, chi2


# DATOS
datos = np.array([
    1.2, 2.1, 2.3, 1.9, 3.0,
    1.5, 2.8, 2.4, 1.7, 2.6,
    2.0, 1.8, 3.1, 2.2, 2.5
])

# nivel de confianza
conf = 0.90


# ESTADÍSTICOS MUESTRALES
# tamaño de muestra
n = len(datos)

# media muestral
x_bar = np.mean(datos)

# varianza muestral (ddof=1 → n-1)
s2 = np.var(datos, ddof=1)

# desviación estándar muestral
s = np.sqrt(s2)


# a) MEDIA CON σ CONOCIDA
# sigma conocida
sigma = 2

# error estándar
se_z = sigma / np.sqrt(n)

# intervalo usando normal
LI_z, LS_z = norm.interval(conf, loc=x_bar, scale=se_z)


# b) MEDIA CON σ DESCONOCIDA
# grados de libertad
df = n - 1

# error estándar
se_t = s / np.sqrt(n)

# intervalo usando t
LI_t, LS_t = t.interval(conf, df, loc=x_bar, scale=se_t)


# c) VARIANZA (χ²)
# intervalo de chi-cuadrada
chi2_low, chi2_high = chi2.interval(conf, df)

# construcción del intervalo para σ² (inversión por asimetría)
LI_var = (df * s2) / chi2_high
LS_var = (df * s2) / chi2_low


# RESULTADOS
print("Estadísticos muestrales:")
print(f"n = {n}")
print(f"Media = {x_bar:.4f}")
print(f"Varianza = {s2:.4f}\n")

print("a) Intervalo para μ (σ conocida):")
print(f"({LI_z:.4f}, {LS_z:.4f})\n")

print("b) Intervalo para μ (σ desconocida):")
print(f"({LI_t:.4f}, {LS_t:.4f})\n")

print("c) Intervalo para σ²:")
print(f"({LI_var:.4f}, {LS_var:.4f})")
```

## **<font color="purple">Ejemplo: </font>**

En un estudio de préstamos a estudiantes, el Departamento de Educación informó que los beneficiarios del fondo Stafford Loan deberían un promedio de 12,658 al recibirse (*USA Today*, 5 de abril de 1995). Suponga que este promedio de deuda se basa en una muestra de 480 préstamos a estudiantes y que la desviación estándar de la población de las deudas al recibirse es 2,000.

##### **<font color="black">Se pide:</font>**

a) Determina un estimado de confianza del 90% del promedio poblacional de la deuda.  

b) Determina un estimado de confianza del 95% del promedio poblacional de la deuda.  

c) Determina un estimado de confianza del 99% del promedio poblacional de la deuda.  

d) Describe lo que sucede con el ancho del intervalo de confianza a medida que se aumenta el nivel de confianza. ¿Parece razonable? Explica tu respuesta.

```python
# INTERVALOS DE CONFIANZA PARA LA MEDIA CON σ CONOCIDA

# importamos librerías necesarias
import numpy as np
from scipy.stats import norm


# DATOS
# media reportada
x_bar = 12658

# tamaño de muestra
n = 480

# desviación estándar poblacional conocida
sigma = 2000


# VALIDACIONES
if n <= 0:
    raise ValueError("n debe ser mayor que 0")

if sigma <= 0:
    raise ValueError("sigma debe ser mayor que 0")


# CÁLCULO DEL ERROR ESTÁNDAR
se = sigma / np.sqrt(n)


# NIVELES DE CONFIANZA
niveles = [0.90, 0.95, 0.99]


# RESULTADOS
print(f"Error estándar: {se:.2f}\n")

for conf in niveles:

    # intervalo usando normal
    LI, LS = norm.interval(conf, loc=x_bar, scale=se)

    # ancho del intervalo
    ancho = LS - LI

    print(f"Nivel de confianza: {int(conf*100)}%")
    print(f"Intervalo: ({LI:.2f}, {LS:.2f})")
    print(f"Ancho del intervalo: {ancho:.2f}\n")


# INTERPRETACIÓN AUTOMÁTICA
print("Interpretación:")
print("A mayor nivel de confianza, el intervalo se vuelve más amplio.")
print("Esto es razonable porque se requiere mayor certeza para capturar el verdadero parámetro.")
```

Hasta ahora, los intervalos de confianza nos han permitido **estimar** un parámetro poblacional dentro de un rango de valores plausibles.

Sin embargo, en muchas situaciones no basta con estimar, sino que se requiere **tomar una decisión concreta** sobre un valor específico.

Por ejemplo:

- ¿Es el promedio realmente igual a cierto valor?  
- ¿Existe evidencia de que ha cambiado?  

En estos casos, en lugar de construir un intervalo, planteamos una afirmación formal y evaluamos si los datos proporcionan evidencia suficiente para rechazarla.

Este cambio nos lleva al estudio de las **pruebas de hipótesis**, donde el objetivo principal es tomar decisiones basadas en evidencia estadística.

## **<font color="navy">Pruebas de Hipótesis</font>**

Una **<font color="crimson">prueba de hipótesis</font>** es un procedimiento estadístico que permite **tomar una decisión** sobre un parámetro poblacional a partir de información muestral.

A diferencia de los intervalos de confianza, aquí no buscamos un rango, sino responder:

> ¿Existe evidencia suficiente para rechazar una afirmación sobre la población?


### **<font color="darkgreen">Estructura de una prueba</font>**

Toda prueba se basa en dos hipótesis:

#### **<font color="purple">Hipótesis nula</font>**

$$
H_0
$$

- Es la afirmación inicial  
- Representa estabilidad o “no cambio”  
- Siempre contiene igualdad  

**Ejemplo:**

$$
H_0: \mu = 50
$$

#### **<font color="purple">Hipótesis alternativa</font>**

$$
H_1 \text{ o } H_a
$$

- Es lo que queremos demostrar  

#### **Tipos:**

##### **Bilateral:**
$$
H_1: \mu \neq 50
$$

##### **Cola derecha:**
$$
H_1: \mu > 50
$$

##### **Cola izquierda:**
$$
H_1: \mu < 50
$$


### **<font color="darkgreen">Nivel de significancia</font>**

$$
\alpha
$$

- Probabilidad de cometer error tipo I  
- Valores comunes:

$$
\alpha = 0.05,\; 0.01,\; 0.10
$$


### **<font color="darkgreen">Estadístico de prueba</font>**

Transforma los datos en una variable con distribución conocida.

#### **<font color="purple">Casos especiales</font>**

##### **Media con $\sigma$ conocida:**

$$
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
$$

##### **Media con $\sigma$ desconocida:**

$$
T = \frac{\bar{X} - \mu_0}{S / \sqrt{n}} \sim t(n-1)
$$


##### **Varianza:**

$$
\chi^2 = \frac{(n-1)S^2}{\sigma_0^2}
$$


##### **Razón de varianzas:**

$$
F = \frac{S_1^2}{S_2^2}
$$


### **<font color="darkgreen">Región crítica</font>**

Es el conjunto de valores donde se **rechaza $H_0$**.

Depende de:

- $\alpha$
- tipo de prueba  


### **<font color="darkgreen">Regla de decisión</font>**

- Si el estadístico cae en la región crítica → **rechazar $H_0$**  
- Si no → **no se rechaza $H_0$**


### **<font color="darkgreen">Valor p</font>**

El valor p es:

> la probabilidad de observar un resultado al menos tan extremo como el obtenido, suponiendo que $H_0$ es verdadera.

#### **<font color="black">Regla</font>**

- Si $p \leq \alpha$ → rechazar $H_0$  
- Si $p > \alpha$ → no rechazar $H_0$


### **<font color="darkgreen">Errores</font>**

#### **<font color="black">Error Tipo I</font>**

Rechazar $H_0$ siendo verdadera  

$$
P(\text{Error Tipo I}) = \alpha
$$

#### **<font color="black">Error Tipo II</font>**

No rechazar $H_0$ siendo falsa  

$$
\beta
$$


### **<font color="darkgreen">Relación con intervalos de confianza</font>**

- Si el valor hipotético está dentro del intervalo → no se rechaza $H_0$  
- Si está fuera → se rechaza  


### **<font color="darkgreen">Pasos de una prueba de hipótesis</font>**

1. Definir $H_0$ y $H_1$  
2. Elegir $\alpha$  
3. Seleccionar el estadístico  
4. Calcular el valor observado  
5. Determinar región crítica o valor p  
6. Tomar decisión  
7. Interpretar  

Muy importante:

- “Rechazar $H_0$” → hay evidencia suficiente en contra  
- “No rechazar $H_0$” → no hay evidencia suficiente  

Nunca se dice “aceptar $H_0$”


## **<font color="purple">Ejemplo: Prueba de Hipótesis para la Media ($\mu$) con $\sigma$ conocida</font>**


Una muestra aleatoria de 100 muertes registradas en Estados Unidos el año pasado reveló una vida promedio de 71.8 años. Si se supone una desviación estándar de la población de 8.9 años, ¿esto parece indicar que la vida media actual es mayor que 70 años? Utilice un nivel de significancia de 0.05.

### **<font color="navy">Solución: </font>**

**Hipótesis:**
$$
H_0: \mu \leq 70
$$

$$
H_1: \mu > 70
$$

Prueba de una cola (derecha)

**Nivel de significancia:**

$$
\alpha = 0.05
$$

**Estadístico de prueba:**

$$
Z = \frac{\bar{X} - \mu_0}{\sigma / \sqrt{n}}
$$

Sustituyendo:

$$
Z = \frac{71.8 - 70}{8.9 / \sqrt{100}} = \frac{1.8}{0.89} \approx 2.02
$$

**Región crítica:**

Para una prueba de una cola:

$$
Z_{0.05} = 1.645
$$

**Regla de decisión:**

$$
\text{Rechazar } H_0 \text{ si } Z > 1.645
$$

**Decisión:**

$$
Z = 2.02 > 1.645
$$

Por lo tanto:

$$
\text{Se rechaza } H_0
$$

**Valor p:**

$$
p \approx 0.0217
$$

$$
p < \alpha = 0.05
$$

Se confirma la decisión de rechazar $H_0$

> Existe evidencia estadística suficiente para afirmar que la vida media actual es mayor que 70 años.

```python
# PRUEBA DE HIPÓTESIS Z (σ CONOCIDA)

# importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# DATOS
x_bar = 71.8     # media muestral
mu0 = 70         # valor hipotético
sigma = 8.9      # desviación estándar poblacional
n = 100          # tamaño de muestra
alpha = 0.05     # nivel de significancia


# VALIDACIONES
if n <= 0:
    raise ValueError("n debe ser mayor que 0")

if sigma <= 0:
    raise ValueError("sigma debe ser mayor que 0")


# CÁLCULOS
# error estándar
se = sigma / np.sqrt(n)

# estadístico Z
Z = (x_bar - mu0) / se

# valor crítico (cola derecha)
z_crit = norm.ppf(1 - alpha)

# valor p (cola derecha)
p_value = 1 - norm.cdf(Z)


# RESULTADOS
print("Resultados de la prueba:\n")
print(f"Estadístico Z: {Z:.4f}")
print(f"Valor crítico: {z_crit:.4f}")
print(f"Valor p: {p_value:.4f}\n")

if Z > z_crit:
    print("Decisión: Se rechaza H0")
else:
    print("Decisión: No se rechaza H0")

if p_value < alpha:
    print("\nInterpretación:")
    print("Existe evidencia suficiente para afirmar que la media es mayor que 70.")
else:
    print("\nInterpretación:")
    print("No hay evidencia suficiente para afirmar que la media es mayor que 70.")


# GRÁFICA

# eje x
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure()

# curva normal
plt.plot(x, y)

# región crítica
x_fill = np.linspace(z_crit, 4, 1000)
y_fill = norm.pdf(x_fill)
plt.fill_between(x_fill, y_fill)

# línea punteada en valor crítico
plt.axvline(z_crit, linestyle='--')

# punto en valor crítico
y_crit = norm.pdf(z_crit)
plt.scatter(z_crit, y_crit)
plt.text(z_crit, y_crit, f"  {z_crit:.2f}")

# punto en Z
y_Z = norm.pdf(Z)
plt.scatter(Z, y_Z)
plt.text(Z, y_Z, f"  Z ≈ {Z:.2f}")

# texto región crítica
plt.text(z_crit + 0.2, max(y)/2, "Región de rechazo")

# línea base
plt.axhline(0)

# estil
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# títulos
plt.title("Prueba Z - Cola derecha")
plt.xlabel("Z")
plt.ylabel("")

plt.show()
```

## **<font color="purple">Ejemplo: Prueba de Hipótesis para la Media ($\mu$) con $\sigma$ desconocida</font>**


El Edison Electric Institute publica cifras del número de kilowatts-hora que gastan anualmente varios aparatos electrodomésticos. Se afirma que una aspiradora gasta un promedio de 46 kilowatts-hora al año. Si una muestra aleatoria de 12 hogares, que se incluye en un estudio planeado, indica que las aspiradoras gastan un promedio de 42 kilowatts-hora al año con una desviación estándar de 11.9 kilowatts-hora, ¿esto sugiere que las aspiradoras gastan, en promedio, menos de 46 kilowatts-hora al año a un nivel de significancia de 0.05? Suponga que la población de kilowatts-hora es normal.

### **<font color="navy">Solución:</font>**

**Hipótesis:**

$$
H_0: \mu \geq 46
$$

$$
H_1: \mu < 46
$$

Prueba de una cola (izquierda)

**Nivel de significancia:**

$$
\alpha = 0.05
$$

**Estadístico de prueba:**

$$
T = \frac{\bar{X} - \mu_0}{S / \sqrt{n}}
$$

Sustituyendo:

$$
T = \frac{42 - 46}{11.9 / \sqrt{12}}
$$

$$
T = \frac{-4}{3.435} \approx -1.16
$$

**Región crítica:**

Grados de libertad:

$$
df = 11
$$

**Valor crítico:**

$$
t_{0.05,11} \approx -1.796
$$

**Regla:**

$$
\text{Rechazar } H_0 \text{ si } T < -1.796
$$

**Decisión:**

$$
T = -1.16 > -1.796
$$

$$
\Rightarrow \text{No se rechaza } H_0
$$


> No existe evidencia suficiente para afirmar que las aspiradoras consumen, en promedio, menos de 46 kilowatts-hora al año.

```python
# PRUEBA DE HIPÓTESIS t (σ DESCONOCIDA) + GRÁFICA DE REGIÓN CRÍTICA

# importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


# DATOS
x_bar = 42       # media muestral
mu0 = 46         # valor hipotético
s = 11.9         # desviación estándar muestral
n = 12           # tamaño de muestra
alpha = 0.05     # nivel de significancia


# VALIDACIONES
if n <= 1:
    raise ValueError("n debe ser mayor que 1")

if s <= 0:
    raise ValueError("s debe ser mayor que 0")


# CÁLCULOS
# grados de libertad
df = n - 1

# error estándar
se = s / np.sqrt(n)

# estadístico t
T = (x_bar - mu0) / se

# valor crítico (cola izquierda)
t_crit = t.ppf(alpha, df)

# valor p (cola izquierda)
p_value = t.cdf(T, df)


# RESULTADOS
print("Resultados de la prueba:\n")
print(f"Estadístico t: {T:.4f}")
print(f"Valor crítico: {t_crit:.4f}")
print(f"Valor p: {p_value:.4f}\n")

if T < t_crit:
    print("Decisión: Se rechaza H0")
else:
    print("Decisión: No se rechaza H0")

if p_value < alpha:
    print("\nInterpretación:")
    print("Existe evidencia suficiente para afirmar que la media es menor que 46.")
else:
    print("\nInterpretación:")
    print("No hay evidencia suficiente para afirmar que la media es menor que 46.")


#  GRÁFICA

# eje x
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

plt.figure()

# curva t
plt.plot(x, y)

# región crítica (cola izquierda)
x_fill = np.linspace(-4, t_crit, 1000)
y_fill = t.pdf(x_fill, df)
plt.fill_between(x_fill, y_fill)

# línea punteada en valor crítico
plt.axvline(t_crit, linestyle='--')

# punto en valor crítico
y_crit = t.pdf(t_crit, df)
plt.scatter(t_crit, y_crit)
plt.text(t_crit, y_crit, f"  {t_crit:.2f}")

# punto en estadístico T
y_T = t.pdf(T, df)
plt.scatter(T, y_T)
plt.text(T, y_T, f"  T ≈ {T:.2f}")

# texto región crítica
plt.text(t_crit - 1.5, max(y)/2, "Región de rechazo")

# línea base
plt.axhline(0)

# estilo limpio
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# títulos
plt.title("Prueba t - Cola izquierda")
plt.xlabel("t")
plt.ylabel("")

plt.show()
```

## **<font color="purple">Ejemplo: Prueba de Hipótesis para Diferencia de Medias ($\mu_1 - \mu_2$)</font>**


Mediciones de resistencia al corte, obtenidas de pruebas de compresión no confinada para dos tipos de suelos, proporcionaron los resultados que se muestran en la siguiente tabla (medidas en toneladas por pie cuadrado). ¿Parecen diferir los suelos con respecto al promedio de resistencia al corte con un nivel de significancia de 1%?


|                | Tipo de suelo I | Tipo de suelo II |
|----------------|----------------|------------------|
| $n$            | 30             | 35               |
| $\bar{y}$      | 1.65           | 1.43             |
| $s$            | 0.26           | 0.22             |

### **<font color="navy">Solución: </font>**

**Hipótesis:**

$$
H_0: \mu_1 = \mu_2
$$

$$
H_1: \mu_1 \neq \mu_2
$$

Prueba bilateral

**Nivel de significancia:**

$$
\alpha = 0.01
$$

**Estadístico de prueba (Welch)**

$$
T = \frac{\bar{y}_1 - \bar{y}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

Sustituyendo:

$$
T = \frac{1.65 - 1.43}{\sqrt{\frac{0.26^2}{30} + \frac{0.22^2}{35}}}
$$

$$
T = \frac{0.22}{\sqrt{0.00225 + 0.00138}} = \frac{0.22}{0.0602} \approx 3.65
$$

**Grados de libertad (Welch):**

$$
df \approx 60
$$

**Región crítica:**

$$
t_{0.005,60} \approx 2.66
$$

Regla:

$$
\text{Rechazar } H_0 \text{ si } |T| > 2.66
$$

**Decisión:**

$$
|T| = 3.65 > 2.66
$$

$$
\Rightarrow \text{Se rechaza } H_0
$$

> Existe evidencia estadística suficiente para afirmar que los dos tipos de suelo difieren en el promedio de resistencia al corte.

```python
# PRUEBA DE HIPÓTESIS PARA DIFERENCIA DE MEDIAS

# importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import t


# DATOS
# muestra 1
n1 = 30
x1 = 1.65
s1 = 0.26

# muestra 2
n2 = 35
x2 = 1.43
s2 = 0.22

# nivel de significancia
alpha = 0.01


# VALIDACIONES
if n1 <= 1 or n2 <= 1:
    raise ValueError("n1 y n2 deben ser mayores que 1")

if s1 <= 0 or s2 <= 0:
    raise ValueError("las desviaciones deben ser positivas")


# CÁLCULOS
# diferencia de medias
diff = x1 - x2

# error estándar (Welch)
se = np.sqrt((s1**2 / n1) + (s2**2 / n2))

# estadístico t
T = diff / se


# grados de libertad (Welch)
num = (s1**2 / n1 + s2**2 / n2)**2
den = ((s1**2 / n1)**2) / (n1 - 1) + ((s2**2 / n2)**2) / (n2 - 1)
df = num / den


# valor crítico (bilateral)
t_crit = t.ppf(1 - alpha/2, df)

# valor p (bilateral)
p_value = 2 * (1 - t.cdf(abs(T), df))


# RESULTADOS
print("Resultados de la prueba:\n")
print(f"Diferencia de medias: {diff:.4f}")
print(f"Estadístico t: {T:.4f}")
print(f"Grados de libertad: {df:.2f}")
print(f"Valor crítico: ±{t_crit:.4f}")
print(f"Valor p: {p_value:.4f}\n")

if abs(T) > t_crit:
    print("Decisión: Se rechaza H0")
else:
    print("Decisión: No se rechaza H0")

if p_value < alpha:
    print("\nInterpretación:")
    print("Existe evidencia suficiente para afirmar que las medias son diferentes.")
else:
    print("\nInterpretación:")
    print("No hay evidencia suficiente para afirmar que las medias son diferentes.")


# GRÁFICA

# eje x
x = np.linspace(-4, 4, 1000)
y = t.pdf(x, df)

plt.figure()

# curva t
plt.plot(x, y)

# regiones críticas (dos colas)
x_fill_left = np.linspace(-4, -t_crit, 1000)
x_fill_right = np.linspace(t_crit, 4, 1000)

plt.fill_between(x_fill_left, t.pdf(x_fill_left, df))
plt.fill_between(x_fill_right, t.pdf(x_fill_right, df))

# líneas críticas
plt.axvline(-t_crit, linestyle='--')
plt.axvline(t_crit, linestyle='--')

# puntos críticos
plt.scatter(-t_crit, t.pdf(-t_crit, df))
plt.text(-t_crit, t.pdf(-t_crit, df), f"  {-t_crit:.2f}")

plt.scatter(t_crit, t.pdf(t_crit, df))
plt.text(t_crit, t.pdf(t_crit, df), f"  {t_crit:.2f}")

# punto del estadístico
y_T = t.pdf(T, df)
plt.scatter(T, y_T)
plt.text(T, y_T, f"  T ≈ {T:.2f}")

# texto regiones
plt.text(-t_crit - 1.2, max(y)/2, "Rechazo")
plt.text(t_crit + 0.2, max(y)/2, "Rechazo")

# línea base
plt.axhline(0)

# estilo limpio
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# títulos
plt.title("Prueba t - Dos colas (Welch)")
plt.xlabel("t")
plt.ylabel("")

plt.show()
```

## **<font color="purple">Ejemplo: Prueba de Hipótesis para Diferencia de Proporciones ($p_1 - p_2$)</font>**


El estado de California está trabajando muy duro para asegurar que todos los estudiantes en edad escolar, cuyo lenguaje materno no sea el inglés, tengan suficiente conocimiento de inglés cuando lleguen al sexto grado. Su avance es supervisado cada año por medio de un examen de desarrollo del idioma inglés en California. Los resultados para dos distritos del sur de California en el año escolar de 2003 se muestran en la siguiente tabla. ¿Estos datos indican una diferencia importante en las proporciones de estudiantes del año escolar 2003 que dominan el inglés en los dos distritos? Use $\alpha = 0.01$.


|                      | Riverside | Palm Springs |
|----------------------|----------|--------------|
| $n$                  | 6124     | 5512         |
| $\hat{p}$ (%)        | 40       | 37           |


### **<font color="navy">Solución: </font>**

**Hipótesis:**

$$
H_0: p_1 = p_2
$$

$$
H_1: p_1 \neq p_2
$$

Prueba bilateral

**Nivel de significancia:**

$$
\alpha = 0.01
$$

**Proporciones:**

$$
\hat{p}_1 = 0.40, \quad \hat{p}_2 = 0.37
$$

***Proporción combinada:**

$$
\hat{p} = \frac{n_1 \hat{p}_1 + n_2 \hat{p}_2}{n_1 + n_2}
$$

$$
\hat{p} = \frac{6124(0.40) + 5512(0.37)}{6124 + 5512} \approx 0.386
$$

**Estadístico de prueba:**

$$
Z = \frac{\hat{p}_1 - \hat{p}_2}{\sqrt{\hat{p}(1-\hat{p})\left(\frac{1}{n_1} + \frac{1}{n_2}\right)}}
$$

Sustituyendo:

$$
Z = \frac{0.40 - 0.37}{\sqrt{0.386(1-0.386)\left(\frac{1}{6124} + \frac{1}{5512}\right)}}
$$

$$
Z \approx \frac{0.03}{0.0090} \approx 3.33
$$

**Región crítica:**

$$
z_{0.005} = 2.576
$$

**Regla:**

$$
\text{Rechazar } H_0 \text{ si } |Z| > 2.576
$$

**Decisión:**

$$
|Z| = 3.33 > 2.576
$$

$$
\Rightarrow \text{Se rechaza } H_0
$$

**Valor p:**

En este caso, como la prueba es **bilateral**, el valor p se calcula como:

$$
p = 2 \cdot P(Z > |3.33|)
$$

Buscando en la distribución normal estándar:

$$
P(Z > 3.33) \approx 0.00043
$$

Entonces:

$$
p = 2 \cdot 0.00043 \approx 0.00086
$$


$$
p = 0.00086 < 0.01
$$

$$
p < 0.01
$$

> Existe evidencia estadística suficiente para afirmar que las proporciones de estudiantes que dominan el inglés son diferentes entre los dos distritos.

```python
# PRUEBA DE HIPÓTESIS PARA DIFERENCIA DE PROPORCIONES

# importamos librerías
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm


# DATOS
n1 = 6124
n2 = 5512

p1 = 0.40
p2 = 0.37

alpha = 0.01


# VALIDACIONES
if n1 <= 0 or n2 <= 0:
    raise ValueError("n1 y n2 deben ser mayores que 0")

if not (0 < p1 < 1 and 0 < p2 < 1):
    raise ValueError("las proporciones deben estar entre 0 y 1")


# CÁLCULOS
# proporción combinada
p_hat = (n1*p1 + n2*p2) / (n1 + n2)

# error estándar
se = np.sqrt(p_hat * (1 - p_hat) * (1/n1 + 1/n2))

# estadístico Z
Z = (p1 - p2) / se

# valor crítico bilateral
z_crit = norm.ppf(1 - alpha/2)

# valor p bilateral
p_value = 2 * (1 - norm.cdf(abs(Z)))


# RESULTADOS
print("Resultados de la prueba:\n")
print(f"Proporción combinada: {p_hat:.4f}")
print(f"Error estándar: {se:.6f}")
print(f"Estadístico Z: {Z:.4f}")
print(f"Valor crítico: ±{z_crit:.4f}")
print(f"Valor p: {p_value:.6f}\n")

if abs(Z) > z_crit:
    print("Decisión: Se rechaza H0")
else:
    print("Decisión: No se rechaza H0")

if p_value < alpha:
    print("\nInterpretación:")
    print("Existe evidencia suficiente para afirmar que las proporciones son diferentes.")
else:
    print("\nInterpretación:")
    print("No hay evidencia suficiente para afirmar que las proporciones son diferentes.")


# GRÁFICA

# eje x
x = np.linspace(-4, 4, 1000)
y = norm.pdf(x)

plt.figure()

# curva normal
plt.plot(x, y)

# regiones críticas (dos colas)
x_fill_left = np.linspace(-4, -z_crit, 1000)
x_fill_right = np.linspace(z_crit, 4, 1000)

plt.fill_between(x_fill_left, norm.pdf(x_fill_left))
plt.fill_between(x_fill_right, norm.pdf(x_fill_right))

# líneas críticas
plt.axvline(-z_crit, linestyle='--')
plt.axvline(z_crit, linestyle='--')

# puntos críticos
plt.scatter(-z_crit, norm.pdf(-z_crit))
plt.text(-z_crit, norm.pdf(-z_crit), f"  {-z_crit:.2f}")

plt.scatter(z_crit, norm.pdf(z_crit))
plt.text(z_crit, norm.pdf(z_crit), f"  {z_crit:.2f}")

# punto del estadístico
y_Z = norm.pdf(Z)
plt.scatter(Z, y_Z)
plt.text(Z, y_Z, f"  Z ≈ {Z:.2f}")

# texto regiones
plt.text(-z_crit - 1.2, max(y)/2, "Rechazo")
plt.text(z_crit + 0.2, max(y)/2, "Rechazo")

# línea base
plt.axhline(0)

# estilo limpio
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)

# títulos
plt.title("Prueba Z - Diferencia de proporciones (dos colas)")
plt.xlabel("Z")
plt.ylabel("")

plt.show()
```
